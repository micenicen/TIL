```python
# text_keyword결측값 대체하기
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from sklearn.preprocessing import OneHotEncoder

apply_train=pd.read_csv('apply_train.csv')
# 지원한 회사를 one-hot encoding으로 표현

# 지원한 회사를 one-hot encoding으로 표현
ohe = OneHotEncoder()
encoded = ohe.fit_transform(apply_train[['recruitment_seq']])

# 다음 줄을 수정합니다.
try:
    column_names = ohe.get_feature_names_out(['recruitment_seq'])
except AttributeError:
    column_names = ohe.get_feature_names(['recruitment_seq'])

encoded_df = pd.DataFrame(encoded.toarray(), columns=column_names)
apply_train_ohe = pd.concat([apply_train[['resume_seq']], encoded_df], axis=1).groupby('resume_seq').sum().reset_index()

similarity = cosine_similarity(apply_train_ohe.iloc[:, 1:])

# 유사도 행렬을 DataFrame으로 변환
similarity_df = pd.DataFrame(similarity, index=apply_train_ohe['resume_seq'], columns=apply_train_ohe['resume_seq'])

# 결측값 대체 함수
def fill_missing_keywords(row):
    if pd.isnull(row['text_keyword']):
        # 유사도가 가장 높은 지원자들을 정렬
        sorted_similar_resumes = similarity_df[row['resume_seq']].sort_values(ascending=False).index.tolist()
        
        # 유사도 순으로 NaN이 아닌 text_keyword 값을 찾기
        for similar_resume in sorted_similar_resumes:
            value = data[data['resume_seq'] == similar_resume]['text_keyword'].values[0]
            if not pd.isnull(value):
                return value
    return row['text_keyword']

# 결측값 대체
data['text_keyword'] = data.apply(fill_missing_keywords, axis=1)
data
```

```python
import pandas as pd

# 유사도 데이터와 지원 내역 데이터를 불러오기
cosine_similarity_df =user_similarity
apply_train = pd.read_csv('apply_train.csv')

def get_recommendations(resume_seq, cosine_similarity_df, apply_train, top_k=5):
    # 지원자와 가장 유사한 다른 지원자들의 인덱스 가져오기.
    similar_applicants = cosine_similarity_df.loc[resume_seq].nlargest(top_k + 1).index[1:].tolist()

    
    # 유사한 지원자들이 지원한 회사 목록.
    recommended_companies = []
    for sim_applicant in similar_applicants:
        companies_applied = apply_train[apply_train['resume_seq'] == sim_applicant]['recruitment_seq'].tolist()
        recommended_companies.extend(companies_applied)
    
    # 중복을 제거하고 상위 5개 회사를 반환합니다.
    unique_recommended = list(set(recommended_companies))
    
    # 만약 5개 미만이라면, 유사도가 다음으로 높은 지원자의 지원 회사를 추가.
    while len(unique_recommended) < top_k:
        additional_applicant = cosine_similarity_df.loc[resume_seq].drop(similar_applicants).idxmax()
        additional_companies = apply_train[apply_train['resume_seq'] == additional_applicant]['recruitment_seq'].tolist()
        unique_recommended.extend(additional_companies)
        similar_applicants.append(additional_applicant)
        unique_recommended = list(set(unique_recommended))
    
    return unique_recommended[:top_k]

recommendations = []

for resume_seq in apply_train['resume_seq'].unique():
    recommended_companies = get_recommendations(resume_seq, cosine_similarity_df, apply_train)
    for company in recommended_companies:
        recommendations.append([resume_seq, company])

recommendation_df = pd.DataFrame(recommendations, columns=['resume_seq', 'recruitment_seq'])

# 결과를 CSV 파일로 저장합니다.
recommendation_df.to_csv('recommendations.csv', index=False)
```

```python
# 앙상블기법이 적용된 코드(참고사항)
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def compute_similarity(data_features):
    similarities = cosine_similarity(data_features)
    return pd.DataFrame(similarities, index=data_features.index, columns=data_features.index)

def recommend_jobs_based_on_similarity(user_id, similarity_matrix, application_data, top_k=5):
    similar_items = similarity_matrix.loc[user_id].sort_values(ascending=False).head(top_k + 1).index[1:]
    applied_jobs = set(application_data[application_data['resume_seq'] == user_id]['job_id'])
    recommended_jobs = [job for job in similar_items if job not in applied_jobs]
    return recommended_jobs[:top_k]

def ensemble_recommendations(user_based_recs, job_based_recs):
    # 여기서는 간단히 두 리스트를 합친 후 중복을 제거합니다.
    # 다양한 결합 방법을 적용할 수 있습니다.
    return list(set(user_based_recs + job_based_recs))

user_features = pd.read_csv('user_features.csv').set_index('resume_seq')
job_features = pd.read_csv('job_features.csv').set_index('job_id')
application_data = pd.read_csv('apply_train.csv')

user_similarity_matrix = compute_similarity(user_features)
job_similarity_matrix = compute_similarity(job_features)

all_recommendations = {}

all_users = application_data['resume_seq'].unique()
for user_id in all_users:
    user_based_recs = recommend_jobs_based_on_similarity(user_id, user_similarity_matrix, application_data)
    job_based_recs = recommend_jobs_based_on_similarity(user_id, job_similarity_matrix, application_data)
    
    final_recommendations = ensemble_recommendations(user_based_recs, job_based_recs)
    all_recommendations[user_id] = final_recommendations

# 결과를 DataFrame 형태로 변환
recommendations_list = [(user, job) for user, jobs in all_recommendations.items() for job in jobs]
recommendations_df = pd.DataFrame(recommendations_list, columns=['resume_seq', 'job_id'])

# 결과 저장
recommendations_df.to_csv('ensemble_recommendations.csv', index=False, encoding='utf-8-sig
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the application data
application_data = pd.read_csv('apply_train.csv')

def get_applied_jobs(user_id, application_data):
    return application_data[application_data['resume_seq'] == user_id]['recruitment_seq'].tolist()

def recommend_jobs_based_on_similarity(user_id, user_features, user_similarity, application_data, top_k_jobs=5):
    similar_jobs = set()

    # Get the user index
    user_idx = user_features.index.get_loc(user_id)

    # Get the jobs that this user applied for
    user_applied_jobs = get_applied_jobs(user_id, application_data)

    # Get similar users to this user
    similar_users = np.argsort(-user_similarity[user_idx])[:top_k_jobs]

    # Collect the jobs that similar users applied for
    for idx in similar_users:
        similar_user_id = user_features.index[idx]
        for job_id in get_applied_jobs(similar_user_id, application_data):
            if job_id not in user_applied_jobs:
                similar_jobs.add(job_id)

    # Return the top k jobs
    return list(similar_jobs)[:top_k_jobs]

def get_all_recommendations(user_features, application_data):
    all_recommendations = {}

    # Compute user similarity only once
    user_similarity = cosine_similarity(user_features)
    all_users = application_data['resume_seq'].unique()
    
    for user_id in all_users:
        recommended_jobs = recommend_jobs_based_on_similarity(user_id, user_features, user_similarity, application_data)
        all_recommendations[user_id] = recommended_jobs
    
    return all_recommendations

all_user_recommendations = get_all_recommendations(user_features, application_data)

# Convert the results into a DataFrame and save to CSV
recommendations_df = pd.DataFrame(all_user_recommendations.items(), columns=['resume_seq', 'recommended_job_id'])
recommendations_df.to_csv('user_recommendations.csv', index=False, encoding='utf-8-sig')
```

```python
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split

# 'apply_train.csv' 데이터를 읽어서 Surprise의 데이터셋 형태로 변환합니다.
ratings_dict = {
    'user': application_data['resume_seq'].tolist(),
    'item': application_data['recruitment_seq'].tolist(),
    'rating': [1] * len(application_data)
}

df = pd.DataFrame(ratings_dict)

# Surprise reader 및 dataset을 생성합니다.
reader = Reader(rating_scale=(0, 1))
data = Dataset.load_from_df(df[['user', 'item', 'rating']], reader)

# 학습 데이터로 사용합니다. (전체 데이터)
trainset = data.build_full_trainset()

# SVD 모델을 학습합니다.
algo = SVD()
algo.fit(trainset)

def get_top_n_recommendations(user_id, n=5):
    all_jobs = set(application_data['recruitment_seq'].unique())
    applied_jobs = set(get_applied_jobs(user_id, application_data))
    
    # 지원하지 않은 회사에 대한 예측을 생성합니다.
    not_applied_jobs = all_jobs - applied_jobs
    predictions = [algo.predict(user_id, job_id) for job_id in not_applied_jobs]
    
    # 예측값 기준 상위 n개의 회사를 반환합니다.
    top_n_jobs = sorted(predictions, key=lambda x: x.est, reverse=True)[:n]
    return [pred.iid for pred in top_n_jobs]

# 모든 사용자에 대한 추천 회사 목록을 생성합니다.
all_users = application_data['resume_seq'].unique()
output = []

for user in all_users:
    recommended_jobs = get_top_n_recommendations(user)
    for job in recommended_jobs:
        output.append({'resume_seq': user, 'recruitment_seq': job})

# DataFrame으로 변환한 후 CSV 파일로 저장합니다.
output_df = pd.DataFrame(output)
output_df.to_csv('user_recommendations.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
import numpy as np


# 사용자-아이템 행렬 초기화
user_item_matrix = pd.pivot_table(apply_train, index='resume_seq', columns='recruitment_seq', fill_value=0, aggfunc='size')

# NaN을 0으로 변환
user_item_matrix.fillna(0, inplace=True)
```

```python
from sklearn.decomposition import TruncatedSVD

# SVD 적용
num_components = 50
svd = TruncatedSVD(n_components=num_components)
user_factors = svd.fit_transform(user_item_matrix)
item_factors = svd.components_.T
```

```python
# 사용자와 아이템 간의 유사도를 계산하여 추천 리스트를 생성하는 함수
def recommend_for_user(user_index, user_factors, item_factors, k=5):
    user_vector = user_factors[user_index, :]
    scores = np.dot(item_factors, user_vector)
    top_k_indices = scores.argsort()[::-1][:k]
    return user_item_matrix.columns[top_k_indices]

# 모든 사용자에 대한 추천 리스트 생성
all_recommendations = []
for user in user_item_matrix.index:
    user_index = user_item_matrix.index.get_loc(user)
    recommendations = recommend_for_user(user_index, user_factors, item_factors)
    for rec in recommendations:
        all_recommendations.append([user, rec])

# DataFrame으로 변환 후 CSV 파일로 저장
recommendation_df = pd.DataFrame(all_recommendations, columns=['resume_seq', 'recruitment_seq'])
recommendation_df.to_csv('recommendations.csv', index=False)
```
