```python
# 사실 배열을 시리즈로 만드는 이유, 시리즈를 데이터프레임으로 만드는 이유부터 이해가 힘들 것이다. 
# 자료는 실생활에서 발생되는 모든 것이다. 이러한 자료(문자, 숫자자료, 불린 등)들을 사람이 처리해왔으나 현재는 기계가 처리한다.
# 기계가 처리하기 위해서는 수치로 변환해야 한다. 
# 발생된 자료를 저장해야 저장된 데이터를 기계가 처리한다. 자료를 저장하는 공간을 '변수'라고 한다. 
# 자료는 크기가 정해져있지 않다. 매우 클 수 있고 매우 작을 수도 있다. 일정하지 않다. 
# 이것을 효율적으로 저장하고 처리하기 위해 공간을 분리했다. 자료의 형태에 따라 여러개의 종류를 만들게 되었는데 이것을 자료형(data type)이라고 한다.
# 문제는 파이썬에서 자료를 저장해서 처리하려고 하다보니 일일히 코딩하고 자료를 처리해야 했다. 
# 그래서 데이터 자료구조를 만들었다. 좀더 효율적인 작업이 필요했다.(파이썬 for문 돌리기보다 배열에서 일괄처리가 훨씬나음)
# 자료구조는 대량의 자료를 처리하기 위해 효율적으로 저장 및 관리하기 위한 방법이다. 
# 파이썬에는 자료구조가 리스트, 튜플, 세트, 딕셔너리가 있다. 자료들을 저장하고 처리하기도 했다. 
# 21세기는 데이터가 많아도 너무 많아졌다. 매일매일 데이터가 쏟아진다. for문으로 써도 느리다. 
# 파이썬은 느리다. 파이썬은 무겁다. 그래서 c언어를 통해 빠른 연산을 하게 한 것이 벡터연산이다. 넘파이가 벡터연산을 기반으로 한다. 
# 넘파이는 언어가 아니다. 패키지 이름이 넘파이이다. 대량의 자료를 빠르게 처리하게 하는 용도이다. 
# 넘파이는 단독으로 쓰일 수 없다. 베이스가 있어야 한다. 파이썬이라는 언어에 종속적이다. 파이썬을 먼저 배운 이유이다. 
# 넘파이는 데이터 특화 도구인 것이다. 수치연산이 목적이다. 행렬연산을 빠르게 한다.
# 판다스는 더 나아가서 데이터 분석에 최적화된 도구이다. 넘파이는 데이터분석용이 아니다. 
# 건물짓는 것으로 비유하자면 파이썬은 나무를 직접 베서 건물을 짓는 모양이고 판다스는 조립식 건물 만들듯이 체계적이게 만드는 것이다.
# 건물을 다 지은다음 입주자에게 보고하는 것을 시각화라고 한다. 
# 데이터분석은 과거부터 현재까지 발생한 데이터를 분석하는 것이다. 그를 위해서 넘파이와 판다스를 쓰는 것이고 기초인 파이썬을 배운 것이다. 
# 그리고 과거의 데이터를 통해 패턴을 찾아서 미래를 예측할 수 있다. 그게 머신러닝, 딥러닝이다.
# 그래서 이러한 점을 이해하고 공부해야 한다. 
# 사이파이, 맷플랏리등 패키지도 많다.
# pip install 패키지명(도구이름)으로 몇가지를 설치했을 것이다. 판다스,넘파이는 다 설치되어 있는 상태이다. 엄밀히 말하면 따로 설치해야 한다. 
# 우리는 데이터에널리틱스, 데이터 사이언티스트가 되기 위한 목적으로 여기에 있다.
```

```python
# iris의 내림차순은 어떻게 하는가? petal_length값을 기준으로 내림차순을 하고싶다고 하자.
iris.groupby(iris.species).get_group('setosa')
```

```python
iris.groupby(iris.species).get_group('setosa').sort_values(by='petal_length', ascending=False)
# sort를 써서 정렬하되 어센딩으로 내림차순으로 설정하였다.
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=["소", "중", "대"]).astype(str)
    #전체그룹을 3개로 나눠서 소중대라는 라벨으 붙인 상태다. 

iris["petal_length_class"] = iris.groupby(iris.species).petal_length.transform(q3cut) # 자체적으로 데이터가 바뀌는 trnhansform함수
iris[["petal_length", "petal_length_class"]].tail(10)
```

```python
def get_top1000(g):
    return g.sort_values('births',ascending=False)[:1000]
# g는 데이터프레임이다. g변수는 5개의 열로 구성되어 있다. 여기서 어느 수치를 기준으로 정렬할 지 정하는 것이다.
# ascending는 오름차순, 내림차순 설정이다. (복습)
```

```python
top1000 = names.groupby(['year','sex']).apply(get_top1000)
top1000
# 성별그룹당 1000이라고 했다. 그런데 262000가 아니다. 사실 1880년대에는 이름종류가 1000건도 안되어서 전부 출력된 것이다.
```

```python
df = boys[boys["year"] == 2010]
df
```

```python
df['prop'].sort_values(ascending=False)
```

```python
prop_cumsum = df['prop'].sort_values(ascending=False).cumsum() #누적합. 오랜만에 본다. 
prop_cumsum
```

```python
prop_cumsum.searchsorted(0.02)
# searchsorted함수이다. 위의 것은 0.02를 찾은 지점이다. 인덱스번호 1번에서 찾았다는 뜻이다. 
```

```python
prop_cumsum.searchsorted(0.03)
# 인덱스번호 3번에서 찾았다는 뜻이다. 
```

```python
prop_cumsum.searchsorted(0.5)
# 인덱스번호116이니 즉 117번째에 0.5가 나왔다. 상위 117명이 차지하는 비율이 50%가 넘는다. 
```
