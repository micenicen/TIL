```python
# 텐서플로 트레이닝 
# 현재는 텐서플로 2버젼이다. 세션이 없이 실행되는 버젼이다. 
W = tf.Variable(tf.random.normal([1])) 
print('initial W = ', W.numpy())
print('===========================================')

# session 생성 없이 즉시 실행 (Eager Execution)
# numpy() 메서드 사용하면 numpy 값을 리턴해줌

for step in range(2):
    W = W + 1.0
    print('step = ', step, ', W = ', W.numpy())
```

```python
# 세션을 생성한 다음 실행하는 것이 일반적이다. 1.x버젼은 그래프를 생성하는 단계와 실행하는 단계가 구분되었다. 
# 텐서플로 1버젼은 그래프를 그리도록 설계가 되어져있다. 
# 
# a--- |
#      c ----d -----e
# b--- |
#
# 문자는 노드, 선을 엣지라고 부른다. 이러한 모양을 텐서플로 그래프라고 부른다. 
# 즉. 연산을 하지 않은 상태이다. 세션이 작동하면서 연산이 되어졌다. 
# 딥러닝기초 파일에서 그래프가 나와있다. 
# 엣지를 다른말로 하면 텐서라고 한다. 원은 노드라고 한다. 연산을 담당한다. 
# 1버젼은 빌딩구조와 실행구조는 별도이다. 
# constant는 상수값 지정이다. Variable은 변수값이다. 
# 과거에는 매우 불친절한 모습으로 인해 많은 사람들이 텐서플로 사용을 포기했었다.
```

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

import numpy as np

print(tf.__version__)
```

```python
x_data = np.array([1, 2, 3, 4, 5, 6])
t_data = np.array([3, 4, 5, 6, 7, 8])
# 모델 구축을 해보자. 
# 케라스에서 시퀸셜을 만들던 것이 우리가 배웠던 방식이다. 
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import mnist

(x_train, t_train), (x_test, t_test) = mnist.load_data()

print('\n train shape = ', x_train.shape, 
      ', train label shape = ', t_train.shape)
print(' test shape = ', x_test.shape, 
      ', test label shape =', t_test.shape)

print('\n train label = ', t_train)  # 학습데이터 정답 출력
print(' test label  = ', t_test)     # 테스트 데이터 정답 출력
```

```python
# 학습데이터에 대한 정답분포를 찍어보자.
label_distribution = np.zeros(10)

for idx in range(len(t_train)):

    label = int(t_train[idx])

    label_distribution[label] = label_distribution[label] + 1


print(label_distribution)
# 비슷하게 분포되어있다. 
```

```python
t_train = tf.keras.utils.to_categorical(t_train, num_classes=10)

t_test = tf.keras.utils.to_categorical(t_test, num_classes=10)
```

```python
x_train = (x_train - 0.0) / (255.0 - 0.0)
x_test = (x_test - 0.0) / (255.0 - 0.0)
# 정규화작업은 항상 비슷하다. 
```

```python
# 모델정의를 하였다. 이거 일반적인 DNN모델인거 같다. 
model = tf.keras.Sequential()

model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))

model.add(tf.keras.layers.Dense(100, activation='relu'))

model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
hist = model.fit(x_train, t_train, epochs=30, validation_split=0.3)
```

```python
(x_train[0])
```

```python
model.evaluate(x_test, t_test)
```

```python
plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

plt.plot(hist.history['loss'], label='train loss')
plt.plot(hist.history['val_loss'], label='validation loss')

plt.legend(loc='best')

plt.show()
```

```python
plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.grid()

plt.plot(hist.history['accuracy'], label='train accuracy')
plt.plot(hist.history['val_accuracy'], label='validation accuracy')

plt.legend(loc='best')

plt.show()
```
