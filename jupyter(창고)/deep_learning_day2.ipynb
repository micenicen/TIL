```python
# 딥러닝 2일차 - 딥러닝 원리
# 11월 말에는 머신러닝, 딥러닝을 연계해서 프로젝트를 할 것이다. 
# 팀 프로젝트를 진행할 것이며 역할분담을 잘 하여야 한다. 
# ui, 모델구성 등 많은 것이 필요하다. 
# 딥러닝, 머신러닝에서 급여를 많이 받는 이유가 이것이다. 어렵다. 
# 인공지능 전공자도 3분의 2가 못따라온다. 그만큼 어렵고 오랜시간이 필요하다. 
# 그렇지만 머신러닝, 딥러닝은 긴 시간이 필요하다. 
# 내공을 쌓으면서 인사이트를 키워나가서 좋은 평가를 받아나가자.
# 일은 매우 재밌다. 잘 공부해서 열심히 활약하자.
# 입상을 못해도 참여하는 것에 의의를 두자. 
# 대학교에서는 논문발표가 일상이다. 준비하는 동안에 스트레스를 많이 받는다.
# 해외논문은 어렵다. 거기에 수식이 잔뜩 나온다. 이걸 이해하고 발표를 해야하는게 부담이다.
# 부담이 적은 방향으로 갔으면 좋겠다. 
# 혼자서 할 수 있는 것은 많지 않다. 서로 부스팅이 필요하다. 
# 스터디는 괴로우면 안된다. 그럴바에 혼자하는게 낫다. 즐겁게 하자. 
# 수학과 영어도 해결하고자 하는 의지를 가지자. 
# AI전문인력은 구하기 쉽지 않다. 
# 최고수준의 전문가 영역으로 가기가 쉽지가 않지만 가면 세상이 재밌어진다.
# 취업에 집중을 해야한다. 통계학, 선형대수도 다 공부해야한다. 길게 해야한다. 공부하면 다 알게된다. 
# 공부가 재밌어지면 그 분야의 전문가가 된다. 
# 논문 하나가 인생뿐만 아니라 세계를 바꿀 수 있는 분야이다. 
# 내일은 조별로 미팅을 할 예정이다. 
# 데이터 리터러시가 필요하다. 주어진 데이터를 보는것 뿐만 아니라 데이터를 읽는 문해능력을 가져야한다.
# 단순하게 숫자가지고 처리하는게 아니다. 도메인에 대한 지식이 필요하다. 
```

```python
def top(x, mx, y, my): 
    d = 0
    for i in range(len(x)):             # 4번 반복을 할 것이다.  
        d += (x[i] - mx) * (y[i] - my)  # x값에 평균을 뺀 것에 y값에 평균을 뺀걸 곱하여 d에 추가한다. 
    return d                            # 이게 분자에 들어간다. 
```

```python
a = dividend / divisor#이게 기울기이다. 
```

```python
b = my - (mx*a)        # 이게 y절편이다. 
```

```python
def predict(x):         # x를 전달받는다. 
    return a * x + b    # 기울기*x, y절편을 각각 리턴한다. 
```

```python
# 이 데이터들을 가지고 선형회귀모델을 만들어서 시각화를 하려고 한다. 
# 각각의 데이터의 예측값과 실제값을 반영하는 선(모델)을 구하고자 한다.
# 그 과정에서 선은 점점 기울어져 코스트를 낮춘다.
# 이번에는 그래디언트 디센트 알고리즘을 알아보자.
# 기울기와 y절편을 각각 초기화한다. 
a = 0
b = 0
# 회귀모델은 직선으로 되어있을 것이다. ax+b 형태이다.이것은 ^y(와이햇)이다.
# 시그마(실제값-와이햇(예측값) )제곱이 바로 mse였다. 이걸 편미분해서 코스트함수에 러닝메이트(a)를 곱한다.
# w를 업데이트 하는 것이다. w*a*미분/(미분*w) 코스트(w,b)
# 단순선형회귀가 아니라 곡선형태의 비선형회귀모델을 만들 것이다. 현재의 알파값은 학습률이다. 
```

```python
# 학습률(러닝메이트)
lr = 0.03
# 반복횟수(에폭)
epochs = 2001
```

```python
# x 값이 총 몇 개인지 셉니다.
n=len(x)                                 # 배치사이즈는 4이다. x는 4개가 존재한다. 

#경사 하강법을 시작합니다.
for i in range(epochs):                  # epoch 수 만큼 반복
    
    y_pred = a * x + b                   # 예측 값을 구하는 식이다. 가설함수이다. 
                                         # x는 데이터다. x는 4개가 나온다. [2,4,6,8]
                                         # 연산은 한번에 수행된다.배치사이즈는 4이다. 1에폭당 4건을 학습한다. 

    error = y - y_pred                   # 실제 값과 비교한 오차를 error로 놓습니다.
                                         # y는 [81,93,91,97]과 같이 있었다. 
                                         # 실제값에 위의 값을 뺀다. 
                                         # 실제에서는 애러를 제곱한 다음 루트를 씌운다. 
    
    a_diff = (2/n) * sum(-x * (error))   # 오차 함수를 a로 편미분한 값입니다. 
                                         
    b_diff = (2/n) * sum(-(error))       # 오차 함수를 b로 편미분한 값입니다. 
    
    a = a - lr * a_diff     # 학습률을 곱해 기존의 a 값을 업데이트합니다.
                            # 코스트 함수를 w로 편미분 한 것이다.기존 기울기에서 빼는 것이다. 
                            # 기울기를 학습률에 따라 계속 개선해나가는 것이다. 
        
    b = b - lr * b_diff     # 학습률을 곱해 기존의 b 값을 업데이트합니다.
                            # 절편도 똑같이 개선을 해나간다. 
    
    if i % 100 == 0:        # 100번 반복될 때마다 현재의 a 값, b 값을 출력합니다.
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))
```

```python
# 이게 경사하강법의 실제 원리이다. 이런 원리로 사용되는 것이다. 
# 에폭이 0일때 경사가 가파르다가 점점 경사가 내려가는 것을 볼 수 있다.
# y값은 일정위치에 수렴한다. 
# 이게 텐서플로다. 이걸 이해하면 경사하강법을 전부 이해하는 것이다. 
print(a)
print(b)
```

```python
# 다층 퍼셉트론
# 퍼셉트론 : 입력값을 입력받아 출력하는 단위. 노드가 이 역할을 하고 있다. 
#            뉴런의 역할과 같다. 인풋을 받고 연산을 해서 다음 뉴런으로 아웃풋을 한다.
# 다층퍼셉트론 : 뇌세포처럼 다중 뉴런에서 인풋을 받고 다중 뉴런으로 아웃풋을 한다. 이게 다층 퍼셉트론이다. 
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    print(x)
    

# 이것은 and & or연산(거짓인가? 참인가?)이다.
# 거짓 거짓, 거짓 참(논리합에서 참), 참 거짓(논리합에서 참), 참 참(논리곱에서 참)
# and와 or이다. 이미 배웠다. 
```

```python
def NAND(x1,x2):
    return MLP(np.array([x1, x2]), w11, b1) # 함수 내부 함수이다.
                                            # [0,0]*[-2,-2]+3 이 첫번째다. 
                                            # 4개를 대입하면  1,1,1,0이 나온다. 

def XOR(x1,x2):
    return AND(NAND(x1, x2),OR(x1,x2))      # 함수의 내부부터 수행이 될 것이다. NAND부터 수행된다. 
                                            # NAND는 3,1,1,-1이 나올 것이고 0,1,1,1을 and로 한것이다. 
                                            # [1,0],[1,1],[1,1][0,1]로 대입되는 것이다.
                                            # 결과적으로 and로 대입하면 중간에 [1,1]인 두 개만 나온다. 

def AND(x1,x2):
    return MLP(np.array([x1, x2]), w2, b3)  # 함수안에서 [0,0]*[1,1] - 1 가 0보다 큰가? 이다.
                                            # 여기서 [1,1]이 인풋이 되지 않는 한 0보다 커질 수 없다. 
                                            


def OR(x1,x2):
    return MLP(np.array([x1, x2]), w12, b2) # 가중치와 바이너리를 추가했다. 
                                            # 첫번째는 0과 2를 곱하고 -1을 뺀 것이다.
                                            # 함수안에서 [0,0]*[2,2] - 1 가 0보다 큰가? 이다. 
                                            # 두번째부터는 1과 2를 곱하는게 적어도 한개 이상이다. 
                                            # 결과값이 1이 출력될 것이다. 
                                            # 함수안에서 [1,0]*[2,2] - 1 == 1 이다. 0보다 크다. 
```

```python
# 가중치와 바이어스
w11 = np.array([-2, -2]) # 가중치 변수가 3개가 있다. 
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3                   # 바이어스(y절편)에 해당하는 변수가 3개이다. 
b2 = -1
b3 = -1

# 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1
```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y=OR(x[0], x[1]) # 0 1 1 1 이 나와야 한다. 
    print(y)

```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y=AND(x[0], x[1]) # 0 0 0 1 이 나와야 한다. 
    print(y)
```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y=XOR(x[0], x[1]) # 0 0 0 1 이 나와야 한다. 
    print(y)
```

```python
# 중요한 것은 퍼셉트론이다. 퍼셉트론 하나로 모든 연산이 해결되었다. 
# 다만 퍼셉트론이 중복해서 작동한다. XOR을 하기 위해서 하나의 선으로 그리기 위해서 노력했지만 되지 않았다.
# 그래서 여러개의 선으로 만들면 가능하다는 것을 증명했다. 
# 이걸 다층 퍼셉트론이라고 한다. 이게 딥러닝이다. 다층 퍼셉트론이 곧 딥러닝이다.
# and나 or같은 것은 단순하게 끝난다. xor는 단순하지 않다. 여러개의 퍼셉트론이 필요하다. 
# 퍼셉트론의 조합으로 새로운 결과를 뽑아내는 것이다. 입력층/히든층이 반복되는 것이다. 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping # 모델 체크포인트, 조기중단 클래스이다. 
import os
import pandas as pd

# 데이터를 입력합니다.
df = pd.read_csv('data/wine.csv', header=None)

# 와인의 속성을 X로 와인의 분류를 y로 저장합니다.
X = df.iloc[:,0:12]
y = df.iloc[:,12]

#학습셋과 테스트셋으로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

# 모델 구조를 설정합니다.
model = Sequential()
model.add(Dense(30,  input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()

#모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 학습이 언제 자동 중단 될지를 설정합니다.
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20)
# 성능이 떨어지는 것을 감지하는 기준은? 모니터링하는 변수는 대체로 val_loss를 쓴다. 20번 참는 것이다.
# 이번건 20번 참는다는 것이다. 

#최적화 모델이 저장될 폴더와 모델의 이름을 정합니다.
modelpath="./data/model/Ch14-4-bestmodel.hdf5"

# 최적화 모델을 업데이트하고 저장합니다.
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=0, save_best_only=True)
# 최적점에만 저장한다는 옵션이다. 이것이 중요하다. 

#모델을 실행합니다.
history=model.fit(X_train, y_train, epochs=2000, batch_size=500, validation_split=0.25, 
                  verbose=1, callbacks=[early_stopping_callback,checkpointer])
# 작동중단, 체크포인터를 저장했다. 
```

```python
model.evaluate(X_test,y_test)
# 이게 딥 러닝이다. 말로만 듣던 딥러닝이다. 
```
