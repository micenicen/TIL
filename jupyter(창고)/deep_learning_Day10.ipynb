```python
# 자연어 처리 attention, transfomer, gpt, 파인튜닝이 이번 주에 할 내용이다.
# 상당히 높은 난이도를 가진다. 교과과정 범위를 넘어가는 단계이다.
# 저번시간 yolo를 통해 오브젝트 디텍팅을 했었다. 
# https://jalammar.github.io/visualizing-neural-machine-translation-mechanics-of-seq2seq-models-with-attention/
# 상당히 정리가 잘 된 블로그이다. 여기서 참고해보자. 
# 2014년 조경현 교수가 같이 작성했던 attention기법을 발표했다. 
```

```python
# sequance to sequance 기법
# 이 이후로 sequance to sequance 기법에 대한 많은 연구가 이루어졌다. 
# 구글 번역기도 2016년 sequance to sequance기법을 사용해왔다. 
# 입력은 단어로 이루어진 시퀀스이다. 차례대로 처리된다.
# 출력으로도 다른 언어로 번역된 시퀀스가 차례대로 나온다. 
# 입력 - 인코더 - 컨텍스트(단어의 뜻 벡터) - 디코더 - 출력 순으로 이루어진다. 
# 컨텍스트는 하나의 벡터 형태다. 여기서 RNN을 사용하게 된다. 
# 컨텍스트 벡터는 실수값이 저장된 벡터이다. 높은 수치값은 밝게 표시되어있다. 
# 워드 임베딩을 통해서 단어들을 벡터공간에 표현할 수 있었다. 
# 디코더 쪽에서도 히든 스테이트를 가지고 있다. 
```

```python
# attention 기법
# 사실 긴 문장을 처리하기에는 적절하지는 않다. 문장을 하나의 벡터로 만든다는 것은 매우 어려운 일이다. 
# 그렇기에 attention매커니즘이 탄생하게 된다.
# sequance to sequance 에서 번역을 할때 디코딩 과정에서 현재 단계에서 가장 중요하게 생각할 수 있는 
# 입력 시퀀스에 포커스를 두자는 의미였다. 어느 입력단어에 비중을 두는지 파악하는것이다. 
# '나는 너를 사랑한다'에서 'i love you'가 있을 경우에 첫 단어는 '나는'에 잡중하고 다음은 '사랑한다'에 집중하는 식이다. 
# 일반 sequance to sequance보다 더 나은 결과를 가져왔다.
# 일반적으로 두 가지 차이점이 존재한다.
# 인코더에서 디코더로 넘어가는 정보량 - 어텐션이 더 많음.
# 디코더의 작업에 인코더가 글자 하나하나마다 관여하는 것이다. 
# 디코더에 end 태그가 나올 때 까지 이 과정이 진행이 된다. 
```

```python
# 입력데이터가 올라간다. 데이터가 199개가 있다고 하자. 
# 입력데이터는 히든레이어로 올라간다. 디코더로 넘어가는 자료는 마지막 히든 스테이트만 올라간다. 
# 입력 문장과 출력 문장은 길이는 다를 수 있다. 
# 재밌는 것은 attention은 기존 sequance to sequance와는 달리 모든 히든 스테이트가 올라간다. 
# 여기서 현재의 단어들에 대한 정보만 올라가는게 아니라. 이전 단어들에 대한 정보도 올라가는 것이 차이점이다.
# 확률의 의미는 가중치이다. 어느 단어에 집중하는지를 말한다. 히든스테이트 값에 가중치를 곱한다. 
# 물론 논문의 핵심은 attention weight의 비중이긴 하다. 여기까지는 생략한다. 
```

```python
# 히든 스테이트는 상태값이다. RNN구조에서 출력되는 결과값이 히든 스테이트라고 한다. 
# 히든 레이어는 일반적인 신경망의 뉴런역할을 하는 값이다. RNN으로 값이 올라가는 곳이 히든 레이어이다. 
# 각각의 RNN 셀들마다 히든 스테이트가 존재한다. 
```

```python
# transfomer
# https://jalammar.github.io/illustrated-transformer/
# 그래도 attention은 LSTM을 기반으로 만들어졌기에 성능에 대해서 RNN의 단점을 그대로 계승한다는 문제가 생겼다. 
# lstm 기반의 sequance to sequance를 따르면서도 어텐션만 구현했다. 
# lstm을 쓰지 않고 attention만 쓰는 것은 어떤가? 라는 발상에서 시작했다. 
# 그래서 RNN을 사용하지 않고 self attention을 이용한 것이다. 
# 인코더, 디코더 내부에서만 자체적인 attention을 수행한다. 
# 이전인코더와 다음 인코더 사이에서 연산을 적용하여 소프트맥스를 적용하는 방식으로 단어의 우선순위를 결정한다. 
# 인코더와 디코더를 하나로 해서 레이어로 정의한다. 
# 어텐션은 병렬로 어텐션을 수행하고 합친다. 
# 한 문장은 인코더 여러개를 통과하는데 통과할때마다 의미가 점점 달라진다. 
# 핵심적인 의미에 해당되는 단어만 남기고 나머지의 비중이 줄어즌다. 
# 결론적으로 attention의 매커니즘을 이용해서 자연어 처리의 역사를 새로 썼다. 
# 트랜스포머 안에는 인코딩도 해야하고 디코딩도 해야힌다.
# 재밌는 것은 트랜스포머 안에는 인코더와 디코더가 여러개가 존재한다. 다만 인코더와 디코더의 갯수는 일치한다.
# 인코더와 디코더는 각각 동일한 구조를 가진다. 차이점은 가중치가 다르다. 웨이트가 전부 다르다. 0
# RNN은 가중치가 각 구조에 동일하게 부여되는 것과 차이가 존재한다.
# 물론 한번에 이해하기는 어렵다. 상당히 최신 논문이다. 논문이해를 완벽하게 하는 것은 어렵다. 
# 인코더의 입력으로 들어오는 구조는 셀프 어텐션(self-attention)이다.
# 인코더에서 하나의 특정 단어를 인코딩하기 위해 다른 단어들과의 관계를 살펴본다. 
# 이후 피드 포워드(feed-foward)에서 출력하는 형식이다. 
# 디코더 내부는 사정이 다른데 셀프 어텐션 - 어텐션 - 피드포워드 순으로 배치된다. 
# -인코더부터 셀프어텐션까지
# 1. 입력단어에 대해 임베딩 알고리즘을 통해 벡터로 변환(크기 512차원으로 변환 - 하이퍼퍼러미터는 직접 수정가능)
# 2. 가장 밑 인코더부터 출발하게 됨.
# 3. 임베딩은 가장 밑에서만 수행
# 4. 인코딩된 것은 다른 인코더로 다시 투입된다.
# 5. 임베딩된 단어에 해당하는 벡터는 인코더 내에서 2개의 서브 레이어에 들어간다. 쿼리벡터와 키벡터에 적재된다. 
# 6. 인코더들과 디코더들은 전부 독립적이다. 따로따로 들어가게 된다. 
# 7. 또한 다양한 단어들에 대해 병렬로 처리한다.
# - 셀프어텐션
# 1. 인코더의 입력벡터, 임베딩벡터에서 3개의 벡터가 만들어진다. 쿼리, 키, 밸류에 곱해진다. 
# 2. 새로운 3개의 벡터들은 기존 임베딩 벡터보다는 작다. 여기있는 3개의 벡터크기는 64차원이다. 이것도 하이퍼퍼러미터이다. 
# 3. 인코더에 대해 어느 단어에 포커스를 맞출지 결정한다. 
# 4. 다른 단어와 관계성은 스코어로 계산한다. 그리고 벡터크기의 제곱근으로 나눈다. (64차원의 제곱근은 8이다.)
# 5. 물론 자기 자신이 가장 높은 점수를 가질 것이다. 
# 6. 각각의 단어들의 위치에 따른 중요도를 체크한다. 의미는 동일하지만 어순이 바뀌는 경우가 존재한다. ()"간다 나는", 나는 간다")
# 7. 재밌는 것은 thinking이 소프트맥스로 88% 나오면 machine은 thinking은 12%가량의 유사도가 나온다. thinking도 machine과 
#  일부 연관이 있다는 것이다. 단어 하나만으로 thinking 단어 하나만으로 설명하기 모자란다는 뜻이다. 여러 단어가 있어야 
#  그 단어의 의미가 보조된다는 것이다. 
# - 출력부
# 1. 벨류벡터에 요소값을 곱한다. 집중을 하고자 하는 단어는 살려두고 관련없는 단어는 점점 제거해간다. 
# 2. 참고로 밸류벡터도 훈련을 통해서 얻어진다. 
# 3. 웨이티드 밸류벡터를 모두 합한다. 그것이 셀프 어텐션 레이어의 출력이 이것이다. 
# 4. 이 결과물을 피드 포워드 신경망으로 보낸다. 이것이 바로 GPT의 인코더 부분 중 일부이다.VERT와 GPT의 기반인 것이다. 
# 5. 이 전체 연산은 행렬로 연산이 수행된다. 상당히 빠르게 수행되는 것이다. 
# 
#  
# 첫 단어는 자체 인코딩 되어져서 첫 인코더로 투입된다. 모든 단어는 병렬로 들아간다. 
# 알고리즘은 심플하지 않다. 맥락을 연관짓는 것을 상당히 어려워한다. 
# self attention은 현재 시점에 처리하고 있는 단어에 다른 연관단어를 연결하여 이해시킨다. 
# 사실 self attention이 트랜스포머의 핵심이다. 
# 
# multi-headed-attention은 해당 논문의 또 다른 핵심이다. 
# 어텐션 계층의 성능을 올리는 또 다른 요소이다. 어텐션헤드가 여러 개가 존재한다는 것이다. 
# 어느 단어든 자기 자신이 가장 높은 점수를 부여한다. 문제는 인칭대명사가 나올때가 존재한다. it, 그것 등이 존재한다.
# 그래서 representation subspaces라는 새로운 공간을 만든다.
# 여러 각도로 문장을 구성하는 관계를 말한다. 즉 단어 사이의 관계를 확인하는 것이다. 
# 처음에는 랜덤한 상태로 있지만 학습을 통해 각각에 대한 공간 행렬값이 수렴이 이루어진다. 인칭대명사에 대한 할당 공간이 만들어진다.
# 결국 복잡한 과정을 거쳐서 하나의 벡터값을 얻어낸다. 
# 
# 다만 사이트 보면 알겠지만 엄청나게 많다. 이게 끝이 아니다. 
# 코사인 유사도가 다시 튀어나온다. 단어의 순서대로 저장해야 한다. 순서도 중요하기 때문이다. 
# 이것을 positional incording(위치 인코딩)이라고 말한다. 수학적인 방법으로 위치를 가볍게 표현하는 것이다. 
# 
```

```python
# 프랑스어, 영어간의 기계번역기를 제작해보자. 
# https://blog.keras.io/a-ten-minute-introduction-to-sequence-to-sequence-learning-in-keras.html
# 시퀀스 투 시퀀스 기반 번역기이다. 
import os
import shutil
import zipfile

import pandas as pd
import tensorflow as tf
import urllib3
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

def download_zip(url, output_path):
    response = requests.get(url, headers=headers, stream=True)                # 리퀘스트를 받아라. 
    if response.status_code == 200:                                            # 서버응답이 정상인가? 비정상이면 4로 시작, 5는 서버오류
        with open(output_path, 'wb') as f:                                     # 아웃풋 패스에 열어라. 
            for chunk in response.iter_content(chunk_size=8192):                # 청크사이즈는 8192사이즈로 설정하여 분할하여 다운로드 받아라. (버퍼)
                f.write(chunk)
        print(f"ZIP file downloaded to {output_path}")
    else:
        print(f"Failed to download. HTTP Response Code: {response.status_code}")

url = "http://www.manythings.org/anki/fra-eng.zip"
output_path = "fra-eng.zip"
download_zip(url, output_path)

path = os.getcwd()
zipfilename = os.path.join(path, output_path)

with zipfile.ZipFile(zipfilename, 'r') as zip_ref:                                  # 저장하라. 
    zip_ref.extractall(path)
# 프랑스어 - 영어번역기를 압축파일로 가져와봤다. 
# 각각의 단어들을 번역한 것을 늘어놓고 동음이의어끼리 묶어놓았다. 
# 참고로 대한민국어도 있다. 상태는 별로 좋진 않다.
```

```python
# 데이터가 구분되어져있는데 \t로 구분되어있는 골때리는 자료이다. 3번째 열부터는 필요없다. 
lines = pd.read_csv('fra.txt', names=['src', 'tar', 'lic'], sep='\t')# 탭 기호로 구분되어야 한다. 
del lines['lic']                                                     # 3번째 열은 날리자.
print('전체 샘플의 개수 :',len(lines))
```

```python
lines = lines.loc[:, 'src':'tar']
lines = lines[0:60000] # 6만개만 저장
lines.sample(10)
# src의 단어가 먼저 올라갈 것이다. 
# 스타트 시그널이 들어가야 한다. 
# 훈련은 교사강요를 기반으로 훈련해야 한다. 
# end가 나와야 끝날 것이다.
```

```python
lines.tar = lines.tar.apply(lambda x : '\t '+ x + ' \n')
lines.sample(10)
# \t는 시작심볼이다. 그리고 \n은 끝 심볼이다. 타깃문장의 앞뒤에 시그널 문자를 추가해야 한다. 
# 토큰의 단위를 문자단위로 한다는 것은 단어가 아닌 문자가 예측되는 것이다. 
# 토큰의 단위를 문장으로 한다면 문장 단위로 예측이되어 나오게 된다. 
# 만약 문장이 있다면 어떤 문자가 왔느냐에 따라 다음 문자를 예측한다는 것이다. 
```

```python
# 문자 집합 구축
src_vocab = set()
for line in lines.src: # 1줄씩 읽음
    for char in line: # 1개의 문자씩 읽음
        src_vocab.add(char)
```

```python
# 타깃열에 대한 문자집합 구축
tar_vocab = set()
for line in lines.tar:
    for char in line:
        tar_vocab.add(char)
```

```python
# 문자단위로도 번역이 가능하고 단어단위로도 가능하다. 사전작업인 전처리를 완료하였다. 
# 프랑스 문장이 길긴 한 모양이다. 
# 훈련할 때와 테스트할 때가 다르다. 교사 강요를 하기 때문이다.
# 기존에 딥 러닝 모델을 만들 때에 시퀸셜 클래스를 생성하고 모델에 add해서 층을 추가했었다. 
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense
from tensorflow.keras.models import Model
import numpy as np
# 이상하게도 시퀸셜이 없다. 대신 input과 MOdel이 존재한다. 
# 케라스에서는 두 가지방식으로 구현을 많이 한다. 지금 것은 낮은 레벨의 설계이고 이것보다 더 상세한 설계도 가능하다. 
# input은 입력 클래스이다. 
```

```python
encoder_inputs = Input(shape=(None, src_vocab_size)) # 80사이즈를 대입했다. 입력되는 문자의 종류(차원)을 말한다. 
# 문자에 대한 길이를 none으로 해주면 문자의 길이는 알아서 판단한다. 
# return_state는 마지막에 나온 글자를 리턴해야한다. 
```

```python
encoder_lstm = LSTM(units=256,return_state=True)# 마지막 글자를 리턴한다는 조건이다. 
# 이것이 인코더 부분이다. 
# 인풋이 여기 안으로 들어와야 한다. 
```

```python
encoder_outputs, state_h, state_c =encoder_lstm(encoder_inputs) # 출력차원은 256차원이다. 앞에 변수를 3개 지정해야 한다. 
# 히든 스테이트, 셀상태 2개(이전 셀의 정보)가 출력된다. 
```

```python
encoder_states = [state_h, state_c] # 인코더의 상태값은 따로 묶어서 저장하자. 이것이 컨텍스트 벡터이다. 
```

```python
# 인코더 설계는 여기까지 이루어졌다. 이제 디코더 설계를 할 차례이다. 
decoder_inputs = Input(shape=(None, tar_vocab_size))
```

```python
decoder_lstm = LSTM(units=256,return_state=True, return_sequences=True)# 디코더는 결과가 나와야 한다. 인코더와 다른 점이다. 
```

```python
decoder_outputs, _,_ = decoder_lstm(decoder_inputs,initial_state=encoder_states) # 디코더 인풋과 타겟문자와 인코더 스테이트를 받아야 한다. 
#
#                            디코더 인풋
#          인코더 스테이트     |
# 인코더  ------------------ 디코더
#          타겟 문자. 
# 
# 이런 모양이 만들어진다. 
# initial state는 초기값이다. 인코더 스테이트를 받아서 출력한다. 
# 그리고 디코더는 은닉상태, 셀 상태정보가 필요없다. 그냥 결과만 나오면 된다.
# 그래서 출력값/ 예측값만 나오면 된다. 그것으로 오차를 찾으면 된다. 
# 출력값은 256차원이다. 인간이 알 수 있는 값은 아니다. 
# 프랑스어로 된 것이 정답이다. 104차원으로 이루어진 언어가 정답인 것이다. 
# 소프트맥스로 예측분류를 해야한다. 
# 이제 dense가 나올 시간이다. 
```

```python
decoder_softmax_layer = Dense(tar_vocab_size, activation='softmax') # 디코더 소프트맥스 레이어는 객체 만들어서 레이어로 만든거다
# add를 하지 않는 것 빼면 동일하다. 풀리커넥티드는 이상없이 이루어진다. 
```

```python
decoder_outputs = decoder_softmax_layer(decoder_outputs)
```

```python
model = Model([encoder_inputs,decoder_inputs],decoder_outputs) # 시퀸셜 방식이 아닌 다른 방식으로 사용했다. 
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy')
```

```python
model.fit(x =[encoder_input,decoder_input],y = decoder_target,batch_size=64,epochs=40,validation_split=0.2)# 모델 핏에는 train x, train y를 입력했었다. 여기는 두 개를 한꺼번에 입력해줘야 한다. 
```

```python
# 이제 입력을 하면 번역문장이 나와야 한다. 
# 번역문장이 그럴 듯 하게 나오는지 확인하는 것도 의미있을 듯 하다. 
# 물론 훈련을 했다고 끝나는 것이 아니다. 출력결과는 디코더 아웃풋으로 나올 것이다. 
# 전부 숫자로 인코딩이 되서 디코더로 들어간다. 
# 문제는 교사강요를 했던 훈련과정은 그렇다고 해도 테스트는 교사강요가 없다.
```

```python
# 테스트 과정은 다음과 같다.
# 1. 번역 대상 입력 문장이 인코더에 들어간다 -> 은닉 /셀 상태 -> 디코더
# 2. 디코더의 입력 시그널이 전달(sos, \t)되어야 한다. 
# 3. 디코더는 다음문자예측을 해야한다. 
# 문제는 교사강요가 없기에 중간에 이상한 단어가 나와버리면 계속 틀려버린다. 
```

```python
# 테스트 과정에서의 디코더 정의 
# 이전 시점의 상태들을 저장하는 텐서
decoder_state_input_h = Input(shape=(256,))# 사실 코드는 변화가 없다. 
decoder_state_input_c = Input(shape=(256,))# 물론 언어가 더 많으면 인풋값이 늘어나야 한다. 
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c] # lstm은 스테이트가 두 곳으로 넘어간다. 다음 lstm과 출력차원 등으로 나온다. 

# 문장의 다음 단어를 예측하기 위해서 초기 상태(initial_state)를 이전 시점의 상태로 사용.
# 뒤의 함수 decode_sequence()에 동작을 구현 예정
decoder_outputs, state_h, state_c = decoder_lstm(decoder_inputs, initial_state=decoder_states_inputs) # 출력으로 해석본, 256,256이 나온다. 
# 디코더 스테이트를 이전에 튀어나온 것을 집어넣는다. 

# 훈련 과정에서와 달리 LSTM의 리턴하는 은닉 상태와 셀 상태를 버리지 않음.
decoder_states = [state_h, state_c]                              
decoder_outputs = decoder_softmax_layer(decoder_outputs)
decoder_model = Model(inputs=[decoder_inputs] + decoder_states_inputs, outputs=[decoder_outputs] + decoder_states)
```

```python
# 테스트과정에서의 인코더 정의 
encoder_model = Model(inputs=encoder_inputs, outputs=encoder_states)
```

```python
index_to_src = dict((i, char) for char, i in src_to_index.items())   # 단어사전 만들기
index_to_tar = dict((i, char) for char, i in tar_to_index.items())
```

```python
def decode_sequence(input_seq):
    # 입력으로부터 인코더의 상태를 얻음
    states_value = encoder_model.predict(input_seq,verbose=0) # 입력을 받아서 출력을 하는 모델

    # <SOS>에 해당하는 원-핫 벡터 생성
    target_seq = np.zeros((1, 1, tar_vocab_size))   # 글자크기 확인
    target_seq[0, 0, tar_to_index['\t']] = 1.       # 시작점 설정

    stop_condition = False                          # 어서 일해라
    decoded_sentence = ""                            # 글자는 아직 없음

    # stop_condition이 True가 될 때까지 루프 반복
    while not stop_condition:
        # 이점 시점의 상태 states_value를 현 시점의 초기 상태로 사용
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value,verbose=0)

        # 예측 결과를 문자로 변환
        sampled_token_index = np.argmax(output_tokens[0, -1, :])                # 최댓값에 해당하는 인덱스를 추출
        sampled_char = index_to_tar[sampled_token_index]                        # 한글자씩 추출해서 정의

        # 현재 시점의 예측 문자를 예측 문장에 추가
        decoded_sentence += sampled_char                                        # 글자를 계속 추가하기

        # <eos>에 도달하거나 최대 길이를 넘으면 중단.
        if (sampled_char == '\n' or                                            # 엔터가 나오거나
            len(decoded_sentence) > max_tar_len):                              # 글자길이 초과하면 
            stop_condition = True                                             # 멈춰

        # 현재 시점의 예측 결과를 다음 시점의 입력으로 사용하기 위해 저장
        target_seq = np.zeros((1, 1, tar_vocab_size))                         # 다시 제작하자고. 
        target_seq[0, 0, sampled_token_index] = 1.                            # 시작점을 이전에 출력한 토큰을 기준으로 잡기

        # 현재 시점의 상태를 다음 시점의 상태로 사용하기 위해 저장
        states_value = [h, c]                                                 # 상태창 재활용
 
    return decoded_sentence                                               # 결국 나가는 단어 만들어라. 
```

```python
# 답 체크하기 
for seq_index in [5,6,100,300,1001]: # 입력 문장의 인덱스
    input_seq = encoder_input[seq_index:seq_index+1]           # 인코더의 문장을 불러오기. 0번 인덱스를 안쓰기에 +1쓰기
    decoded_sentence = decode_sequence(input_seq)              # 함수를 적용해서 번역 돌려버리기. 
    print(35 * "-")
    print('입력 문장:', lines.src[seq_index])
    print('정답 문장:', lines.tar[seq_index][2:len(lines.tar[seq_index])-1]) # '\t'와 '\n'을 빼고 출력 : 
    print('번역 문장:', decoded_sentence[1:len(decoded_sentence)-1]) # '\n'을 빼고 출력 : 번역문에는 시작문자를 뺐기 때문
```
