```python
import pandas as pd
import numpy as np

dictional = pd.read_csv('쿵쿵따사전.csv',encoding='utf-8')
myungsa1 = dictional[dictional['part']=='명사']
# 1. 명사의 길이가 최대 4글자인 것 선택
myungsa2 = myungsa1[myungsa1['word'].apply(lambda x: len(x) <= 4)]
myungsa2 = myungsa2[myungsa1['word'].apply(lambda x: len(x) > 1)]
# 2. 글자로만 이루어진 명사 선택 (정규표현식 사용)
myungsa3 = myungsa2[myungsa2['word'].str.match('^[가-힣]+$')]
myungsa3 = myungsa3.drop('id',axis=1)
myungsa3.reset_index(drop=True,inplace=True)  
myungsa3 = myungsa3.drop('part',axis=1)
front = myungsa3['word'].str[0]
end = myungsa3['word'].str[-1]
front = front.rename('front')
end = end.rename('end')
dictiona = pd.concat([myungsa3,front,end],axis=1)
```

```python
import random                                                                # 랜덤모듈
import time

difficuly = int(input('난이도를 입력하세요(초급 20, 중급 50, 고급 100 권장) : '))
if difficuly <=20:
    timeout_seconds = 20
elif difficuly <=50:
    timeout_seconds = 15
elif difficuly <= 100:
    timeout_seconds = 10
elif difficuly > 100:
    timeout_seconds = 5

def make_word_chain(nouns_df, length=3):                                     # 워드체인 만들기, 길이 2
    # 초기 단어 선택
    start_word = random.choice(nouns_df['word'])                             
    current_word = start_word
    chain = [current_word]
    print(f'선공단어 : {chain[0]}')
    print('쿵쿵따')

    for _ in range(length - 1):                                              # 인간 공격
        # 현재 단어의 마지막 글자
        last_letter = current_word[-1]
        start_time = time.time()
        user_attack = input()
        if len(user_attack) > 5 or len(user_attack) < 2:
            print("길이 부적절! 당신의 패배입니다.")
            break
        elif (time.time() - start_time) >  timeout_seconds:
            print("시간초과! 당신의 패배입니다.") 
            break            
        elif user_attack[0] != last_letter:
            print("앞단어 불일치! 당신의 패배입니다.") 
            break
#         elif user_attack.strip() not in dictiona['word'].str.strip().values:
#             print("없는 단어! 당신의 패배입니다.") 
#             break
        elif user_attack in chain:
            print("단어 재사용! 당신의 패배입니다.")
            break
        else:
            chain.append(user_attack)
            current_word = user_attack
            print('쿵쿵따')
                                                                             # 컴퓨터 공격 
        # 현재 단어의 마지막 글자
        last_letter = current_word[-1]
        
        # 해당 글자로 시작하는 단어 중에서 랜덤 선택
        next_word_candidates = nouns_df[nouns_df['front'] == last_letter]
        next_word_candidates.reset_index(drop=True,inplace=True)

        if len(next_word_candidates) >= 3:
            next_word = random.choice(next_word_candidates['word'])
            chain.append(next_word)
            current_word = next_word
            print(current_word)
            print('쿵쿵따')
    
        else:
            # 해당 글자로 시작하는 단어가 3개 미만이면 중단
            print('당신이 승리하였습니다.  종료되었습니다')
            break

    return chain

# 끝말잇기 생성 (예시로 길이 5)
word_chain = make_word_chain(dictiona, length=difficuly)
print(word_chain)
```
