```python
# 디션트리 - 마케팅을 위한 디션트리
# 의사결정트리라고도 한다. 
# 질문을 던지고 답변을 찾아가는 과정에서 의사결정 트리와 같은 것을 구성해서 답을 찾아내도록 하는 것이다.
# 예를 들면 날씨. 기온과 같은 변수에 따라서 야구경기가 열릴 가능성을 어떻게 계산하는지가 그 예시이다. 
# 또다른 것으로 서울 지역(구)의 경도와 위도 정보를 학습하여 임의로 입력된 지역(동)을 강동, 강서, 강남, 강북으로 분류하는 것이다.
# 데이터들이 레이블에 존재할 것이다. 그 레이블은 정답의 특징들이 담겨있다.
# 질문을 던질때 어떤 질문을 던져야 후보 답이 되는 범위가 가장 좁아지는지, 가장 좋은 질문을 하기 위해 노력할 것이다.
# 예를 들면 100이라는 숫자 중 정답이 있을 경우 좋은 질문을 던져서 효율적으로 처리해야 한다.
# 50을 선택하는게 일반적이지만 10을 선택할 수도 있다. 다만 운이 매우 많이 따라야 한다.
# 엔트로피와 지니계수가 여기에 활용된다. 
```

```python
# 의사결정 트리 
# 어떤 항목에 대한 관측값과 목표 값을 연결시켜주는 예측모델로서 결정 트리를 사용하는 머신러닝 방법.
# 쉽게 말하면 아키네이터이다. 결정적인 질문으로 답을 찾아간다.
# 엔트로피는 질문으로 인해서 딱 2 부류로 나눠질 경우 엔트로피가 높다고 한다.
# 하지만 질문한 후에 답이 전부 나온 상황에서는 엔트로피가 0인 상태이다.
# 즉 질문이전 - 질문이후의 엔트로피가 0인 경우는 질문을 했을 때 엔트로피가 높으나 답이 다 결정되었으므로 엔트로피가 0이라고 한다. 
# 하지만 답이 정해진 순간 엔트로피는 낮다고 한다. 이러한 엔트로피를 낮게 만드는 방법이 존재한다.
# 의사결정 트리는 이러한 엔트로피를 낮추는 질문이 핵심이다. 질문의 수준이 간단하면서도 높아야 한다.
```

```python
# 엔트로피
# 불확실성을 수치적으로 표현한 값이다.
# 정보이득은 질문 이전의 엔트로피에서 질문후의 엔트로피를 뺀 값이다. 즉 불확실성이 줄어든 정도이다.
# 엔트로피는 p1-logp1 + p2-logp2로 계산한다. 
# 로그는 딥 러닝할때도 그렇고 필요하다. 로그수식은 필요하다. 
# 일반적으로 언급이 안되어있으면 로그의 밑수가 2인 경우가 많다. 
# 지수함수의 반대는 로그함수이다. 시그모리드 함수를 배울때 코스트함수가 나올 것이다. 여기서 로그가 다시 나올 것이다.
```

```python
# 지니계수
# 얼마나 많은 데이터들이 섞여있는지를 나타내는 불확실성을 의미한다. 0~0.5사이의 값을 가진다.
# 불확실성이 0이라는 뜻은 확실하다는 것을 의미하며 같은 종류에 해당하는 데이터들끼리 존자한다고 한다.
# 지니계수는 작으면 작을수록 불순도가 낮아진다.
# 0.5일때가 데이터가 정확히 반반씩 섞여있는 경우이다.
# CART는 의사결정 나무 알고리즘이다. 불순도를 측정할 때 사용한다. 
# 엔트로피보다 속도가 빠른 장점이 있다.
# 지니계수가 시험볼때는 단답식으로 많이 나온다. 
```

```python
# 의사결정트리는 다중 분류에도 탁월한 성능을 보인다. 
# 다만 언더피팅과 오버피팅은 경계해야 한다. 특징 하나만으로 분류를 하거나 특화된 데이터로만 분류를 하는 식으로 가면 난감하다. 
# 언더피팅은 변수가 너무 적은 상태고 오버피팅은 변수가 너무 많아서 테스트값에서 너무 과적합인 상태이다. 
# 특징의 개수가 너무 많아졌다. 오버피팅을 한다면 테스트 데이터에대해서 정확도가 떨어져있을 가능성이 높다. 일반화가 안된 것이다. 
# 트리를 만들더라도 질문을 너무 많이 던지는 것도 별로 안좋다. 세세한 질문은 더이상 질문할 게 없을정도로 완벽한 질문지가 완성된다.
# 너무 많은 질문을 하면 언젠가 엔트로피는 0이 되겠지만 그건 오버피팅이다. 
# prune은 가지치기라는 뜻이다. pruning을 하도록 모델을 설계해야 한다. 즉 특징을 가지치듯 없애는 것이다. 오버피팅을 방지하기 위함이다.
# max_depse라는 설정이 가지치기를 위한 설정이다.
```

```python
# 의사결정트리의 장점은 수학적인 지식이 없어도 결과를 해석하고 이해하기 쉽다.
# 수치 데이터 및 범주 데이터에 모두 사용 가능하다.
# 정확도가 높은 편이다.
# 하지만 과적합이 자주 나오고 최적의 솔루션을 보장하지 않아서 자주 사용되지 않는다.
# 오버피팅문제가 항상 따라오기 때문에 이로 인해 자주 사용되지 않는 것이다. 구시대적인 기술이기도 하다. 데이터에 따라 다르긴 하나 한계가 명확하다.
# 다음에 앙상블/랜덤포레스트등을 하게 될 것이다. 앙상블 기법이 상당히 좋다. 앙상블은 의사결정 트리를 개선한 기법이다. 
```

```python
#어제 했던 것은 dbscan clustering을 배웠다. 
# kmeans를 사용하고 종료시점(더이삼 점들의 조정이 없을 경우)에 끝나는 것을 배웠다.
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
# 오랜만의 붓꽃데이터이다.
```

```python
iris.keys()# 아이리스 열 인덱스의 이름들이다.
iris.target# 아이리스 타겟들의 이름이다. 사실상 정답이다. 
```

```python
labels =  pd.DataFrame(iris.target)
data = pd.DataFrame(iris.data)
data
```

```python
labels.columns=['labels']
```

```python
data.columns=['Sepal length','Sepal width','Petal length','Petal width']
```

```python
# 꽃받침과 꽃잎의 길이들이 나눠진다. 
pd.concat([data.labels],axis=1)
```

```python
data=pd.concat([data, labels], axis=1)
data
```

```python
feature = data[['Sepal length','Sepal width']]
feature
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = KMeans(n_clusters = 3)# 클러스터를 바꿔서 보는게 가능하다. 
model.fit(feature)# 모델을 생성한다. 데이터의 특징이 반영되도록 모델을 제작되어야 할텐데..
```

```python
pred = pd.DataFrame(model.predict(feature))# feature자리에는 뉴 데이터가 와야한다. 문제는 모의고사 시험지가 실제 시험지와 같은 문제와 같다.
# 모델의 성능이 나쁠 수 밖에 없다. 
pred.columns = ['predict']
# 물론 데이터가 150건밖에 안되기 때문에 쉽지 않을 것이다. 
```

```python
res = pd.concat([feature,pred], axis=1) # 답을 옆에 추가시켰다. 기존에 알고있는 레이블과는 다를 것이다. 
res
```

```python
q1 = data['age'].quantile(0.25) #1사분위수이다. 
q1
```

```python
q3 = data['age'].quantile(0.75) # 3사분위수이다. 
q3# 위의 두 개를 뺀 것이 iqr이다. 
```

```python
iqr = q3-q1
iqr # 이렇게 iqr이 출력되었다. 
```

```python
q1-1.5*iqr #아웃라이어 최소조건
```

```python
q3+1.5*iqr # 아웃라이어 최대조건
```

```python
df = data[((data['age']>(q1-1.5*iqr)) & (data['age'] < (q3+1.5*iqr)))]#나이가 정상범위안에 있는가?
df
# 아웃라이어를 제외한 데이터프레임을 구성했다. 
```

```python
wcss = [] # within cluster sum of square 로 클러스터 데이터들이 뭉쳐있는지 퍼져있는지를 확인한 값을 담는 리스트이다. 
# 값이 적을수록 많이 뭉쳐있다는 뜻이고 값이 클수록 멀리 떨어져있다는 뜻이다. 클러스터링이 잘 되어있다고 보기 어렵다고 할 수 있다. 
for i in range(1,20): #19번 반복
    kmeans = KMeans(n_clusters=i,init='k-means++' ,max_iter=300,random_state=0) 
    #max는 300번이 넘어가도록 멈추지 않으면 강제로 끝낸다는 뜻이다. 랜덤값은 0으로 고정한다는 뜻이다.  
    kmeans.fit(scaled_feature)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,20),wcss) 
plt.title('THe Elbow Curve') #엘보우 그래프
plt.xlabel('Number of Clusters')
plt.ylabel("WCSS")
plt.show()
# 적절한 k를 찾는게 어렵다. 여기서 팔꿈치를 찾아야 한다. 무한정 k를 늘리는게 아니라 적정값을 찾아야 한다.
# 급격하게 경사가 떨어지다가 완만해지기 시작해지는 시점이다. 밑의 그래프를 기준으로 하면 5~7 사이의 값이 적당하다고 본다. 
```
