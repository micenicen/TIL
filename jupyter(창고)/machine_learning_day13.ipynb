```python
c = df['vote_average'].mean()
c
# 평균이다. 평점의 평균치이다. 
```

```python
m = df['vote_count'].quantile(0.9)
m
# 일정 이상만큼의 votecount를 기준하자. 즉 영화가 차트에 실리려면 전체영화중에 90%이상의 득표수가 있어야만 추천을 해 준다는 것이다.
# 영화 차트에 실리기 위한 최소득표수이다. 
```

```python
q_movies = df.loc[df['vote_count']>=m]
q_movies.shape # 최소득표에 해당되는 영화만 추천목록에 담았다.
```

```python
def weighted_rating(x, m=m, c=c): # 먼저 q_movies를 전달할 것이다. m은 기존의 1838이고 c는 6점일 것이다. 
    v = x['vote_count'] # 해당영화의 득표수이다.
    R = x['vote_average'] # 해당영화의 평점이다. 
    return (v/(v+m) * R) + (m/(m+v) * c)
```

```python
# for문을 써서 반복할 수 있긴 하다. 하지만 시간이 매우 많이 걸린다.
# 이제는 for문보다는 apply문을 쓰자. 
# 데이터 규모가 커질수록 apply를 더 많이 쓰게 될 것이다. 
q_movies['score'] = q_movies.apply(weighted_rating,axis=1)
# 매우 빠르게 끝난다. for문은 안쓰인다. 
```

```python
# 선형대수에서 pca를 어떻게 풀어쓰는가?
# 데이터가 있을때 데이터의 공분산 행렬을 구한다. - > 고유값으로 분해한다. -> 고유벡터가 나온다. -> 고유벡터를 선형변환한다.
# 고유벡터 : pca의 주성분벡터로 입력데이터의 가장 큰 방향을 고유벡터라고 한다. 
# 고유값 : 고유벡터의 크기, 데이터분산을 말한다.
# 과정을 기억하자.
# 1. 원본데이터의 공분산 행렬 구하기
# 2. 공분산 행렬의 고유값과 고유벡터를 구한다.
# 3. 고유값이 가장 큰 고유벡터를 추출한다.(축소하고자 하는 차원의 수만큼)
# 4. 고유값이 가장 큰 고유벡터를 이용하여 원본데이터를 변환한다. (차원축소한다.)
```

```python
# 차원축소의 기술
from sklearn.datasets import load_iris

iris = load_iris()

columns = ['sepal_length','sepal_width','petal_length','petal_width']
iris_df = pd.DataFrame(data = iris.data, columns = columns)
iris_df["target"] = iris.target

iris_df.head() # 우리가 아는 그 붓꽃 데이터 맞다. 데이터는 4차원이다.
```

```python
import matplotlib.pyplot as plt
markers = ["^", "s", "o"]

# 0:setosa, 1:versicolor, 2:virginica
for i, marker in enumerate(markers):
    x_axis_data = iris_df[iris_df['target']==i]['sepal_length']
    y_axis_data = iris_df[iris_df['target']==i]['sepal_width']
    plt.scatter(x_axis_data, y_axis_data, marker=marker, label=iris.target_names[i])

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
# x축은 sepal length, y축은 sepal width로 표현된 것이다. 단순하다. 
# 분류기가 잘 분류를 못하는 경우이다. 2차원 데이터로 하다보니 한계가 나온다.
# 문제는 분류기로 선을 그려서 분류하면 초록과 주황은 분류가 많이 어려울 것이다.
# 4차원 데이터인데 2차원으로 분류한 상태라서 성능이 별로다. 세토사만 잘 구분된다.
```

```python
from sklearn.preprocessing import StandardScaler

iris_f_scaled = StandardScaler().fit_transform(iris_df.iloc[:,:-1])
iris_f_scaled
# 우리는 차원축소를 하려고 한다. 2차원으로 차원축소를 하고자 한다.
# 단순하게 픽쳐셀렉션으로는 안될 것이다. 
# 새로운 두 개의 축을 찾아서 분류기를 만들어보자.
```

```python
# pca알고리즘은 알고리즘이 수행하기 위해 스케일링을 해줘야 잘 작동한다.
# 4차원을 2차원으로 줄여보자.
#from sklearn.decomposition import TruncatedSVD 이건 SVD알고리즘이다. 
from sklearn.decomposition import PCA # 오늘은 이걸 써보자.
```

```python
pca = PCA(n_components=2) # 차원을 2차원으로 바꾼다는 뜻이다. 
```

```python
pca.fit(iris_f_scaled)
```

```python
iris_pca = pca.transform(iris_f_scaled)
```
