```python
# 앙상블 기법
# 앙상블 기법의 특징과 작동 원리를 설명할 수 있다.
# 랜덤 포레스트 알고리즘을 실제 데이터에 적용하여 머신러닝을 수행할 수 있다.
# 보팅 앙상블 기법을 실제 데이터에 적용하여 머신러닝을 수행할 수 있다. 
# (의사결정 트리)디션트리 기법들은 분류결과가 이렇게 나온 이유를 설명하는게 쉬운 편이었다. 구현도 쉬운 편이었다.
# 하지만 디션트리 기법은 오버피팅 문제가 항상 따라다니는 단점이 발생했다. 즉 과적합 문제가 항상 발생했다.
# 모델의 성능을 기대하기도 어렵고 오버피팅 가능성도 높기 때문에 현재에는 잘 쓰이지 않는다.
# 앙상블은 기존의 디션트리 기법의 과적합 문제에 대해 취약하기 때문에 과적합 문제를 해결하기 위해 나온 방법이다. 
```

```python
# 앙상블 기법 그러면 여러개의 어떤 모델이 있는 것이다. 여러개의 모델을 만들고 모델들이 모여서 투표를 통하여 다수결로 판단한다던지 예측한다.
# 마치 도전골든벨 같은 것이다. 굉장히 똑똑한 사람이 맨 마지막에 나오게 되는데 찬스를 쓰면 비행기를 날리는 장면이 있다. 그 장면과 비슷하다.
# 아주 똑똑한 모델 하나보다도 덜 똑똑한 모델 여러개를 사용해서 판단하는 것은 어떤가? 이게 핵심이다.
# 쉽게 말하면 집단지성의 힘을 이용하는 것과 비슷하다. 강하지 않는 여러 개의 모델들이 투표를 해서 판단하면 더 강력한 모델이 만들어진다는 것이 바탕된다.
# 일반화된 모델들을 만들 수 있다는 장점이 존재한다. 
```

```python
# 앙상블 기법은 여러개의 분류를 생성하고 그 예측을 결합하여 더욱 정확한 예측을 도출하는 기법이다.
# 강력한 하나의 모델을 사용하는 것 보다 약한 모델을 여러 개를 조합하여 사용하여 강한 모델을 생성한다는 것이다. 
# 만들때도 모델들이 서로 독립적으로 만들었을때가 성능이 좋은 편이다. 
# 약한 모델 여러개를 만들 때 한 가지 알고리즘으로 만드는 것이 아니라 여러 개의 모델을 이용해서 만드는 것이다. 
# 분류가 약한 여러개를 쓰면 오류가 많이 발생할 것이다. 하지만 오류가 나지 않은 쪽을 사용하는 방식으로 해결할 수도 있어서 유용하다.
```

```python
# 보팅
# 즉 다수결 투표다. 하드보팅과 소프트 보팅을 나뉜다.
# 하드보팅은 다수결 원칙을 적용하는 것을 말한다. 예측된 결과들 중에서 다수의 분류기가 예측한 결과를 최종 결과값으로 정한다.
# 모델이 여러개가 있을 때 의사결정 트리들이 모델의 수만큼 만들어진다. 
# 만약 의사결정트리가 6개일때 이 중 4개 이상이 나오면 해당 항목이 다수결로 결정된다. 
# 소프트보딩은 분류별로 합한 확률을 합해서 평균을 매긴다. 확률이 가장 높은 레이블을 최종 투표값으로 정하는 것이다. 
# 하드 보팅은 결과값만 참조한다. 소프트보팅은 분류결과 각각에 대해서 정답일 확률을 출력하는 것이다. 
# 확률을 전부 더한 다음에 합계를 보는 식으로 운영하는게 일반적이다.
# 단순하게 보기에는 소프트 보팅이 더 좋은 결과를 보이는 것 처럼 보인다.
```

```python
# 하드보팅과 소프트보팅의 예측 결과가 다르게 나오는 경우
# 사실 소프트보팅은 점수로 계산하고 하드보팅은 결과값으로만 계산하기 때문에 둘의 결과가 다를 수 있다.
# 이런 오류가 발생할 가능성이 존재한다. 실제로 많이 발생한다. 일반적으로 이럴 경우 소프트보팅으로 하는 경우가 많다.
```

```python
# 배깅
# 부트스트랩(bootstrap)과 어그리게이팅(aggregating)을 합친 단어가 배깅이다. 
# 샘플(표본)을 여러번 추출하는데 중복을 허용한다. 
# 훈련 데이터셋에서 랜덤하게 추출하여 모델을 학습하는 것이다. 
# 대표적으로 랜덤 포레스트가 있다. 딥러닝 등장 이전까지 독보적인 알고리즘이었다. 
# 지금도 랜덤포레스트를 많이 사용하며 높은 정확도를 가진다. 알고리즘 모르겠으면 랜덤 포레스트 쓰라고 할 정도로 정확도가 높다. 
# 다른 것으로는 서포트 벡터머신(SVM)이 있다. 이것도 성능이 상당히 좋다. 
# 반의어로 페이스팅이 있다. 페이스팅은 중복을 허용하지 않는다. 
```

```python
# 보팅과 배깅의 차이점
# 보팅은 데이터셋에서 일관적으로 뽑아낸다. 디션트리, knn, logistic regression등 각각의 모델을 사용하여 투표하는 것이다. 
# 배깅은 하나의 모델에서 샘플을 여러개를 뽑아서 투표하는 방식이다. 즉 중복을 허용한 샘플링을 하는 것이 핵심이다. 
# 배깅은 중복된 데이터들이 여기저기 있을 수 있다. 일부 랜덤 데이터를 가지고 샘플데이터를 만들고 같은 모델 알고리즘으로 계산하는 것이다.
# 즉 여러 모델을 쓰는 방식이 보팅, 여러 랜덤 데이터를 쓰는 방식이 배깅이다. 
```

```python
# 실무에서는 어떤 알고리즘을 쓰는지 판단해야한다. 데이터의 양에 따라서도 다르다는 뜻이다. 
# 초창기에는 데이터가 많지 않았고 이 시기에는 머신러닝을 사용했다. 데이터가 적으면 머신러닝을 사용한다. 
# 데이터가 엄청나게 많아진다면 딥 러닝을 생각해야 한다. 데이터가 가지고 있는 다양한 패턴들을 딥 러닝 모델로 찾고 구현하겠다는 것이다.
# 사실 데이터가 얼마 있지도 않은 상태에서 딥 러닝을 하는 것은 상당한 뻘짓이다. 
# 딥 러닝을 한다는 것은 데이터의 패턴과 특징이 매우 복잡하다는 것을 전제로 한다. 차원수가 많다보니 신경망의 구조를 복잡하게 설계해야 할 필요가 있을때 쓴다.
# 간단한 그런 문제들은 딥 러닝을 사용하지 않는다. 딥러닝은 데이터가 적은 상태에서 하지 않는다. 성능도 별로다.
# 적은 데이터에 대해서는 머신러닝을 사용하자. 
```

```python
# 앙상블
# 앙상블은 여러 개의 의사결정을 합쳐서 높은 성능을 가져다주는 머신러닝 기법이다. 
# 앙상블 기법의 핵심은 성능이 떨어지는 약한 분류기를 여러개를 만들어서 병렬/ 직렬 연결하여 강한 분류기를 만들어내는 방법이다. 
# 샘플을 여러번 추출하여 의사결정 트리를 여러개 만들어서 집계하여 판단하는 것이다. 
# 여러개의 나무들이 모여 숲을 형성하는데 각각의 나무들이 만들어질때(디션트리) 사용된 데이터가 랜덤으로 샘플링된 데이터이다.
# 사용된 데이터들이 전부 다를 것이고 중복되기도 할 것이다. 그래서 모델을 만들면 모델이 더 안정적이고 정확한 모델이 만들어진다는 것이다. 
# 랜덤포레스트를 사용하기 위해서 설정해야하는 것이 있다.
# 1. 나무를 몇 그루를 쓸 것인가? : 디션트리의 갯수를 설정해야한다.
# 2. 각각의 나무의 깊이(뎁스)를 얼마로 할 것인가? 
# 이런 것들이 하이퍼 퍼러미터로 존재한다.
# 부트스트랩 : 오리지널 데이터에서 랜덤으로 추출하는 과정
# 어그리게이팅 : 분류된 랜덤 데이터를 사용하여 모델을 가동함
# 배깅 : 어그리게이팅이 된 결과를 앙상블 분류기로 가져왐
```

```python
# 물론 데이터가 한 쪽으로 쏠릴 수 있다. 
# 모델에 대한 약간의 편향(bias)이 있으면 모델의 성능이 올라가고 과적합을 해결하는 것도 가능해진다. 
# 약간의 편향은 데이터가 중복되어서 특정 데이터가 약간 많은 것을 의미한다. 편향(bias) = 중복허용을 의미한다.
# 모델이 출력한 예측값의 특징, 모습, 형태 등 이와 같은 것을 이해하고자 할때 사용한다. 편향과 분산이다. 
# 페이스팅은 편향성이 없다. 중복허용이 되지 않기 때문이다.
# 편향성을 약간 가지게 해주면 상대적으로 분산이 낮아진다. 
```

```python
# 부트스트랩 
# 데이터 샘플링 시 편향을 높임으로서 분산이 높은 모델의 과대적합을 줄이는 효과를 가져온다. 
# 각 데이터는 전체 데이터 n개보다 적게 학습하되 중복을 허용하여 편향성을 높인다. 
# 보통 데이터의 크기는 전체에서 60~70%가량의 데이터를 사용한다. 
# 결과는 4개가 나와있다.
# 1. 고편향(bias), 저분산(variance) = 정답만 찍어낸 상태 
# 2. 고편향(bias), 고분산(variance) = 정답 주변에 흩어진 형태
# 3. 저편향(bias), 저분산(variance) = 오답주위에 뭉쳐있는 상태
# 4. 저편향(bias), 고분산(variance) = 딴곳에서 흩어진 형태 
```

```python
# 어그리게이팅
# 여러 분류 모델이 예측한 값들을 조합해서 하나의 결론을 도출하는 과정이다.
# 결론은 투표로 결정한다.
```

```python
# 랜덤포레스트
# 다수의 결정 트리들을 배깅해서 예측을 실행하는 앙상블 기법이다.
# 각 노드에 주어진 데이터를 샘플링해서 일부 데이터를 제외한 채 최적의 특징을 찾아 트리를 분기한다.
# 모델의 편향성을 증가시켜 과대적합의 위험을 감소시킨다.
# 다양한 분야에서 좋은 성능을 보여준다. 
```

```python
# 머신러닝 서적을 보면 수식이 많이 나온다. 
# 혼자 공부하는 머신러닝/ 딥러닝에 263페이지에 있다. 배깅, 스테킹이 없는 것이 아쉽다. 
# 보통 논문을 보면 수식이 많이 나온다 그 기호들은 다음과 같다.

# x : 데이터를 의미한다. 
# F(x) : 정답을 의미한다. y로 표현되기도 한다. 
# F^(x) : 예측값을 의미한다. y^라고 표현되기도 한다. 
# E[ ] : 기대치, 평균을 의미한다. 
# E[F^(x)] 정답 예측값의 기댓값이다. 
```

```python
# 편향(bias)은 정답과 예측값 사이에 얼마나 떨어져있는지를 나타낸다. 즉 정답과 예측값 사이의 거리를 말한다. 
# 거리는 전부 양수로 표현되어야 한다. 음수로 표현되면 안된다. 그래서 제곱을 한다. 
# 제곱은 음수를 제곱할 뿐 아니라 패널티도 더 주는 효과를 준다. 
# 수식에서 등장하는 기호를 바탕으로 편향을 수식으로 표현하면 다음과 같다.
# (E[F^(x)]-F(x))**2 
# (예측값의 기댓값-정답)제곱 으로 해석된다.
# 분산(variance). 즉 예측값들의 평균과 예측값들간의 거리의 차이를 평균내서 제곱하면 된다. 예측값들 간에 흩어진 정도를 나타낸다. 
# E[(f^(x) - E[F^(x)])**2]
# 예측값들마다 차이가 있는 것들을 제곱해서 평균을 낸 것이다. 
# 편향과 분산은 머신러닝 모델을 가지고 예측값과 얼마나 차이가 나느냐. 얼마나 모여있느냐를 표현한다. 
# 편향과 분산은 모델이 어느정도 복잡한가 하는 복잡성을 나타낼 수 있다. 
```

```python
# 실제 데이터의 특성을 잘 반영하고 있어야지 예측결과가 그럴싸하게 나온다. 
# 적당한 편향과 적당한 분산이 이루어져야 한다. 
# 새로운 데이터에 대해 정확도가 높아야 한다. 과적합된 자료는 새로운 데에터에 답을 제대로 하지 못한다. 
# 모델은 예측값을 말한다. 실제 데이터와 차이가 생길 수 있다. 
# 예측값과 실제 데이터사이의 차이가 너무 크면 편향이 높다고 본다. 
```

```python
# 훈련은 적정한 수준에서 끝내야 한다. 그걸 알아내는 방법 중 하나는 그래프를 그려보는 것이다.
# 편향과 분산은 반비례관계(trade off)에 있다. 편향과 분산은 반대방향으로 이동한다. 
# 편향과 분산을 적당한 위치에서 멈추는 방법이 필요할 것이다. 
# 데이터를 훈련하다보면 모델에 대한 복잡도를 이야기하는데 당연히 올라간다.
# 분산은 훈련하는 동안 낮아지다가 다시 올라가는 경향이 있다. 즉 다시 올라가는 턴어라운드 지점의 위치가 멈추는 곳이다. 
# 자동으로 중단을 하는 것도 가능하다. 
```

```python
# 앙상블 기법에는 4가지가 존재한다.
# 1. 보팅
# 2. 배깅
# 3. 부스팅
# 4. 스태킹
# 랜덤 포레스트 배깅 이용 : 트리들이 서로 다른 특성을 갖게 되어서 일반화 성능을 향상시킬 수 있다. 
# 중복을 허용하여 같은 알고리즘을 사용해서 모델링을 한다. 
```

```python
# 부스팅 
# 케글이나 공모전에 가면 상위 랭크된 상당수가 부스팅을 잘 사용한 케이스이다. 
# 부스팅은 학습 결과가 나오면 나름대로 맞춘것도 있고 틀린것도 있는데 
# 틀린것들에 대해 높은 가중치를 둬서 높은 가중치가 부여된 데이터가 다시 샘플링이 되어 훈련에 참여하도록 유도하는 기법이다. 
# 즉 틀린것들을 지속적으로 반복하여 정확성을 개선하는 것이다.
# 동일한 알고리즘을 순차적으로 학습하여 가중투표를 통해 예측값을 결정하는 것이다. 
# 데이터를 샘플링해서 모델을 만든다. 틀린것들에 대해서는 더 높은 가중치를 줘서 다음에 더 가중치를 두는 것이다.
# 쉽게 말하면 오답노트이다. 
# 부스팅을 하면 순차적으로 과정을 반복한다. 그 모델에서 틀린것을 가중치를 줘서 다시 모델링 하는 것을 반복한다.
# 부스팅의 특징은 크게 시퀸셜(sequential)이 존재한다. 
# 배깅은 병렬로 학습했다. 이와 다르게 부스팅에서는 순차적으로 학습한다. 
# 부스팅이라는 것도 결국 학습하는 모델을 여러 개 연결해서 샘플링되고 잘 되도록 유도한다. 훈련을 거듭할수록 더 강력한 모델이 만들어진다.
# 앞단에 있는 모델을 보완해가면서 이후에 모델이 완성이 되는 방식이다. 
```

```python
# 부스팅의 종류
# 에이다 부스팅
# 어댑티드 부스팅
# 그레디언트 부스팅의 하위분류
# - XG 부스트
# - light GBM
# - GBM
```

```python
# 에이다 부스팅 
# 잘 분류된 것은 그대로 두고 잘못된 것을 가중치를 높여서 우선순위를 부여한다.
# 정답값만 참조하여 가중치를 부여하여 분류한다.
# 에이다 부스트 기법은 상당히 많이 사용되었다. 
```

```python
# 어댑티드 부스팅 
# 잘 분류된 것은 비율을 낮추고 잘못 분류된 것에 가중치를 높이는 방식이다. 
# 즉 정답값과 오답값을 다 참조하는 것이다. 
```

```python
# 부스팅의 단계
# 샘플1 - 학습 알고리즘 - 분류기1 - 가중치 업데이트
# 샘플2 -  학습 알고리즘 - 분류기2 - 가중치 업데이트 (반복)
# 전체 분류기 참조 - 최종 - 학습데이터 적용
# 최종결정할때 분류기들에 대해 동일한 영향력으로 최종결정을 하는게 아니다. 
# 만약 분류기를 만들었는데 첫 번째 분류기부터 분류를 진행해나가면서 학습이 잘된 의사결정 트리에는 더 높은 가중치를 주는 것이다.
# 즉 가장 잘 학습된 데이터에 더 큰 영향을 미치도록 한다.
```

```python
# 부스팅 예제
# 인물사진 안에 있는 인물을 보고 남자 또는 여자로 분류하는 의사결정 트리를 부스팅한다고 하자.
# 테스트 결과를 통해 남자사진 데이터가 오답이 높을 경우 남자학습 데이터를 보강하고 다시 학습을 시작.
# 공모전을 나갈 때 잘 모르겠다면 라이트 GBM을 쓰라는 말이 있다. 상위 10%로 들기 매우 쉽다.
```

```python
# 가중투표 - 하드보팅 : 분류기의 성능에 따라 가중치를 반영하여 투표결과를 조정한다.
# 단순 동일한 비중으로 분류하면 마지막 분류기 정확도가 높다고 해도 다른 것들이 똥이면 엉뚱한 답이 나올 수 있다. 
# 만약 남자로 분류한 값이 0.4, 0.5로 나오고 여자로 분류한 값이 0.95로 나올 경우 하드보팅할 경우 여자로 분류한다. 0.9<0.95이기 때문이다.
# 가중투표 - 소프트보팅 : 단일 예측값이 아닌 모든 분류값에 대한 확률에 가중치를 곱한 값으로 최종 결론을 도출한다.
```

```python
# 랜덤포레스트를 만들때의 주요 파라미터는 다음과 같다.
# 1. n_estimator : 모델에서 사용할 트리 개수를 몇개로 할 것인가? 디폴트가 100가량이다. 
#보통 홀수로 트리개수를 준다. 동률을 제거하기 위함이다.
# 2. max_features : 최적의 분할을 위해 고려할 피처의 최대 개수이다. 기본으로는 제곱근이다. 
# 기본은 제곱근으로 설정되었다. 10개를 예로 들면 3개를 추출해서 트리를 만든다는 것이다. 이와 같이 피쳐의 갯수를 설정할 수 있다. 인트값과 소수값을 줄 수 있다. 
# 전체 피쳐중 몇개를 사용하는지 확인하는 것이다. 과적합을 방지하기 위함이다. 
# 로그2 옵션, 결측값, 숫자, 소수 옵션이 존재한다. 
# 3. max_depth : 트리의 최대 깊이 
# 과적합에 빠지지 않도록 설정하는 용도이다. 만약 아무것도 설정 안하면 자동으로 오버피팅까지 된다. 
# 4. min_samples_leaf :  리프 노드가 되기 위한 최소 샘플 데이터 개수
# 트리는 노드와 노드를 잇는 간선(엣지)으로 구성된다. 나뭇가지 역할이다. 뿌리의 노드는 루트노드라고 한다.
# 리프노드는 잎이 되는 노드이다. 즉 가장 마지막에 있는 잎에 해당하는 노드로 종단노드(터미널 노드)라고 한다.
# 리프노드는 더 이상 아래로 뻣어나가지 않는다. 자식노드가 없는 것이다. 자식이 있는 부분은 인터널 로드라고 한다. 
# 5. min_samples_split : 자식노드를 갖기 위한 최소한의 데이터 개수
# 파생하지 않는다는 말은 아직 데이터가 분리되지 않는다는 뜻이다.
# 그 때 데이터가 최소 갯수가 안되면 나누지를 않는다. 
# 이는 오버피팅을 제어하기 위한 카테고리들이다. 값이 너무 낮으면 오버피팅이 된다. 
# ----------------------------------------------------------------------------
# 여기서부터는 잘 안한다. 
# max_leaf_node : 리프노드 갯수를 제한한다. 리프노드 갯수가 여기에 걸리면 바로 중단된다.  기초는 none이다. 과적합 제어용이다. 
# random_state : 속성을 랜덤하게 선택하는데 만약 랜덤시드를 고정하고 싶다면 사용한다. 
# class_weight : 가중치를 줘서 랜덤하게 추출할 때 어떤 것이 더 추출되도록 하는지 설정하는 것이다. 원하는 것을 더 뽑아내고 싶을때 쓴다.
```

```python
def names(train, test):
    for i in [train, test]: # 트레인 데이터와 테스트 데이터가 넘어간다. 데이터 두개는 리스트가 될 것이다. 
        i['Name_Len'] = i['Name'].apply(lambda x: len(x)) 
        i['Name_Title'] = i['Name'].apply(lambda x: x.split(',')[1]).apply(lambda x: x.split()[0])# 타이틀 추출해서 집어넣기 
        del i['Name'] #기존 네임을 삭제한다. 
    return train, test
```

```python
def age_impute(train, test):
    for i in [train, test]:
        i['Age_Null_Flag'] = i['Age'].apply(lambda x: 1 if pd.isnull(x) else 0) # 나이가 기재되어있으면 1 아니면 0으로 대입한다. 플래그 역할을 수행한다. 
        data = train.groupby(['Name_Title', 'Pclass'])['Age'] # 네임타이틀과 pclass를 기준으로 그룹화했다. data에 담아내었다. 
        i['Age'] = data.transform(lambda x: x.fillna(x.mean())) # x의 평균으로nan값을 채웠다.
    return train, test
```

```python
def fam_size(train, test):
    for i in [train, test]:
        i['Fam_Size'] = np.where((i['SibSp']+i['Parch']) == 0 , 'Solo', # 조건문이다. 둘을 더했을 때 0이면 솔로라고 표시한다. 
                           np.where((i['SibSp']+i['Parch']) <= 3,'Nuclear', 'Big'))# 이건 다중조건이다. 솔로가 아니면 3명이하 핵가족, 3명이상 대가족으로 나눈다. 
        del i['SibSp'] 
        del i['Parch']
    return train, test
```

```python
def ticket_grouped(train, test):
    for i in [train, test]:
        i['Ticket_Lett'] = i['Ticket'].apply(lambda x: str(x)[0]) # 티켓의 첫번째 문자를 추출하는 것이다.
        i['Ticket_Lett'] = i['Ticket_Lett'].apply(lambda x: str(x)) # 추출한 단어를 문자로 변환한다. 
        i['Ticket_Lett'] = np.where((i['Ticket_Lett']).isin(['1', '2', '3', 'S', 'P', 'C', 'A']), i['Ticket_Lett'],# where은 if문과 같다. 참이면 위. 아닐때는 아래이다. 
                                   np.where((i['Ticket_Lett']).isin(['W', '4', '7', '6', 'L', '5', '8']),
                                            'Low_ticket', 'Other_ticket'))
        i['Ticket_Len'] = i['Ticket'].apply(lambda x: len(x)) #티켓 길이를 대입한다. 
        del i['Ticket']
    return train, test
```

```python
def cabin_num(train, test):
    for i in [train, test]:
        i['Cabin_num1'] = i['Cabin'].apply(lambda x: str(x).split(' ')[-1][1:])
        i['Cabin_num1'].replace('an', np.NaN, inplace = True)
        i['Cabin_num1'] = i['Cabin_num1'].apply(lambda x: int(x) if not pd.isnull(x) and x != '' else np.NaN)
        i['Cabin_num'] = pd.qcut(train['Cabin_num1'],3)
    train = pd.concat((train, pd.get_dummies(train['Cabin_num'], prefix = 'Cabin_num')), axis = 1)
    test = pd.concat((test, pd.get_dummies(test['Cabin_num'], prefix = 'Cabin_num')), axis = 1)
    del train['Cabin_num']
    del test['Cabin_num']
    del train['Cabin_num1']
    del test['Cabin_num1']
    return train, test # 캐빈_넘을 추가하기 위한 함수이다. 아까전에 추가가 안된 모양이다. 
```

```python
def cabin(train, test):
    for i in [train, test]:
        i['Cabin_Letter'] = i['Cabin'].apply(lambda x: str(x)[0]) # 문자타입 0번 추출하고 기존의 케빈열을 제거했다. 
        del i['Cabin']
    return train, test
```

```python
def drop(train, test, bye = ['PassengerId']): # 패신저 아이디를 제거하는 것이다. 먼저 바이는 패신저 아이디로 설정했다. 
    for i in [train, test]:
        for z in bye: 
            del i[z]# 패신저 아이디를 전부 제거하였다. 
    return train, test
```

```python
def dummies(train, test, columns = ['Pclass', 'Sex', 'Embarked', 'Ticket_Lett', 'Cabin_Letter', 'Name_Title', 'Fam_Size']):
    for column in columns:
        train[column] = train[column].apply(lambda x: str(x))#열값을 주었다.
        test[column] = test[column].apply(lambda x: str(x))
        good_cols = [column+'_'+i for i in train[column].unique() if i in test[column].unique()] # 유니크한것들이 테스트유니크에도 있다면. 컬럼명 뒤에 언더바로 연결하라.
        train = pd.concat((train, pd.get_dummies(train[column], prefix = column)[good_cols]), axis = 1) # 더미함수로 나눠버리자. 
        test = pd.concat((test, pd.get_dummies(test[column], prefix = column)[good_cols]), axis = 1)
        del train[column]
        del test[column]
    return train, test
```

```python
train, test = dummies(train, test, columns = ['Pclass', 'Sex', 'Embarked', 'Ticket_Lett',
                                                                     'Cabin_Letter', 'Name_Title', 'Fam_Size'])
```
