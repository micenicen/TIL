```python
* 미니 프로젝트명 : 곡물가격의 상관관계 / 타 원자재와 곡물가격의 상관관계
---------------------------------------------------------------
* 개요
- 세계의 곡물시장은 미국을 중심으로 움직인다. 미국의 선물거래소를 통해서 거래된다.
- 미국에서 생산하는 콩, 밀, 옥수수등은 전 세계에 영향을 끼친다.
- 곡물가격과 여러 요소의 상관관계를 파악하여 곡물가격에 가장 영향을 많이 끼치는 요소를 파악하는 것을 목적으로 한다. 
* 계획
- 기간은 2014~2023년까지의 곡물의 수확량(금년은 예상량)을 토대로 곡물의 수확량/ 소비량 등을 종합하고 가격과의 상관관계를 파악한다.
- 각 국가의 수출량과 소비량을 확인하고 수출국과 수입국의 곡물 가격 상관관계를 파악한다.
- 석유의 가격과 곡물의 가격의 상관성을 확인한다. 
* 데이터확보
- 데이터는 3개의 사이트에서 종합했다.
- https://kr.tradingview.com/chart/                   이 사이트는 차트 사이트로 가격데이터를 크롤링하는 사이트이다. 
- https://www.igc.int/en/markets/marketinfo-sd.aspx   이 사이트는 국제곡물위원회로 전 세계 곡물의 생산량과 소비량을 확인한다.
- https://www.cmegroup.com/ko/education.html          이 사이트는 시카고거래소사이트로 곡물, 원자재의 정보를 확인한다. 
* 데이터 정리 계획
1. 국제곡물위원회 데이터로 곡물의 수확량 / 소비량 / 가격 등의 상관관계를 확인한다.
2. 국제곡물위원회 데이터로 곡물의 수입 / 수출량 / 가격 등의 상관관계를 확인한다. 
3. 가격데이터를 이용하여 석유의 가격과 곡물의 가격의 상관관계를 확인한다.
* 데이터 가공 계획
- 파이차트로 수확량 / 수출입 차트를 기재한다.
- 곡물간의 가격데이터와 석유의 상관관계는 히트맵으로 확인한다.
```

```python
import pandas as pd
import os

grain = ['곡물전체','밀','보리','쌀','옥수수','콩']
country = ['나이지리아','남아프리카','러시아','멕시코','미국','베트남','브라질','사우디아라비아','아르헨티나','영국','우크라이나','유럽연합',
          '이집트','인도','인도네시아','일본','중국','카자흐스탄','캐나다','태국','터키','필리핀','한국','호주']
dictDf = {}  

for i in grain:
    for j in country:
        filename = f"grain_database\\{i}-{j}.csv"
        df_name = f"{i}_{j}"
        
        if os.path.exists(filename):
            dictDf [df_name] = pd.read_csv(filename,index_col= 'm. tons',na_values="-")
            dictDf [df_name].columns.names = ["seasons"]
# 파일 1차 가공 - df에 딕셔너리 형태로 자료 전부 저장하기
```

```python
# 시즌별로 세부분류
dictseason = {}
for i in grain:
    for j in country:
        if f'{i}_{j}' in dictDf: # 킷값이 존재하는지 검사하기.
            length1 = len(dictDf[f'{i}_{j}'].columns)
            for k in range(length1):
                season_name= dictDf[f'{i}_{j}'].columns[k]
                dictseason[f"{i}_{j}_{season_name}"] = dictDf[f'{i}_{j}'][season_name]
dictseason
```

```python
# 카텍고리별로 세부분류
dictcategory = {}
for i in grain:
    for j in country:
        if f'{i}_{j}' in dictDf: # 킷값이 존재하는지 검사하기.
            length2 = len(dictDf[f'{i}_{j}'].index)
            for l in range(length2):
                category_name= dictDf[f'{i}_{j}'].index[l]
                dictcategory[f"{i}_{j}_{category_name}"] = dictDf[f'{i}_{j}'].loc[category_name]
dictcategory
```
