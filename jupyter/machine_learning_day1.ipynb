```python
# -- d 두번 눌러 삭제됨 --
# RNN은 시계열적인 특성을 가질때 적절한 방식이다. 
# 시계열 특징을 가진 데이터라고 한다면 RNN구조를 먼저 떠올릴 수 있다. 
# 문장을 먼저 생성하고 생성한 문장을 음성으로 바꿔서 챗을 하는데 문장을 생성하는 과정은 순서가 존재한다. 
# 전후관계가 있다는 것은 단어들 사이에 시계열적 특성이 있다는 뜻이다. 
# RNN은 미래를 예측하거나 텍스트를 생성하는 과정에서도 많이 사용되어진다. 
# 일반적으로 영상, 이미지와 같은 비전분야같은 네트워크분야는 CNN을 사용하며 텍스트, 미래예측은 RNN이라는 딥러닝 구조를 사용한다.
# 물론 반대로 CNN으로 미래예측에 사용하거나 RNN으로 비전분야에 사용할 수도 있다. 다만 바닐라로서는 어려울 것이다. 
# 이미지 식별을 위한 논문들이 많이 출현하였다.
```

```python
# RNN이 가지고있는 한계점이 존재하는데 그 문제점을 개선한 것이 LSTM이다. 
# RNN의 문제점은 앞의 문장을 통해서 뒤의 문장을 예측할 수 없다는 것이다. 장기 의존성문제이다.
# 인간은  "나는 프랑스에서 자랐다.(중략)나는 000어를 사용한다." 라는 말을 들으면 당연히 프랑스어라고 대답할 것이다. 
# 하지만 기계는 중략된 곳에서 찾는 동안에 답을 이상하게 낼 수 있다. 
# 이를 해결한 구조가 LSTM이다. 이 뒤로도 후속버젼이 계속 나온다. 
# 분류모델, GPT, 파인튜닝등을 해볼 것이다. 이후에 우리는 챗봇을 만들 수 있게 될 것이다. 
# GRU는 한국에서 제안된 사항으로 장기의존성을 다소 해결한 알고리즘이다. RNN의 장기의존성을 어느정도 해결한 구조이다. 
# 두 번째 프로젝트를 할 때에는 팀 프로젝트로 진행될 것이다. 공부해야 할 내용이 상당히 많다. 
# 되도록 오전에 수업을 하고 오후에 프로젝트를 하는 것으로 하자. 
```
