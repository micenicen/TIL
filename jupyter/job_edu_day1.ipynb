```python
# 프로젝트
# 남들이 만들어놓은 것을 따라 만드는 것은 의미없다.
# 내가 이것을 서비스를 통해서 할 수 있을것 같다는 느낌의 프로젝트를 수행해야 한다.
# 현실성이 매우 중요하다. 회사에서는 필요없는 것을 만들지 않는다. 
# 비즈니스는 비영리적으로라도 쓸 때 의미가 있다. 
# 대회를 나가는 것도 비슷한 이유이다. 
# 대회를 나가는 것은 포트폴리오의 질을 높여준다. 
# 문제점, 개선점, 현실적, 비즈니스 등등을 고려한다. 
# 일반적으로 Mnist는 글자를 판변할 때 사용했다. mnist로 자동화하는 것이다. 
# 심사위원들은 "만들었으니까 쓰실 건가요?"라고 물어볼것이다. 
# 본인이 만든 것은 본인이 쓸 수 있어야 한다. 만든 사람도 안쓰는 것은 누구도 쓰지 않는다. 
# 기술을 배우는 이유는 쓸모 있는 것을 만들기 위해서 배우는 것이다. 
# 이외에는 포트폴리오는 쓸모가 없다. 인터넷에 코드가 널려있어서 복사 붙여넣기하면 끝난다. 
# 창업을 하는 사람은 더더욱 잘 해야 한다. 투자를 받기 위해서 이를 정확히 이해해야 한다. 
# 케라스를 쓸 수 있는 이유는 만든 이유를 전부 정의했다. 하지만 우리는 그냥 만든다. 깃헙을 하더라도 친절하게 잘 써주는게 중요하다.
# 나만 보기 위한 것은 깃헙에 올리면 끝이다.  하지만 내가 짠 코드를 한달 뒤에 내가 알아볼 수 있는지는 별개이다. 
# 나를 위해서라도 설명을 남겨둬야 한다. 주석은 그런 역할이다. 대부분은 한달만 지나도 내 코드를 못알아본다.
```

```python
# 나의 PPT는 나를 위한 PPT였다. 남들에게 설명할때 쉽지가 않다. 
# 매우 어려운 주제의 PPT였다. 
# PPT를 할 때에 흰 배경에 검정색을 적는 것도 괜찮을 것이다. 
# PPT를 쓰는 것의 핵심은 글을 쓰지 않기 위해서 PPT를 만든다. 
# 설명을 들으면서 읽는 것은 불가하다. 발표자들은 글만 보거나 발표만 듣는다. 즉 하나만 취한다.
# 사람들이 볼 때 눈에 잘 안들어온다. 요약을 활용하는 것이 중요하다. 
# 글보다는 그림이다. 그게 PPT에 치명적이다. 
# 발표스킬과 관련이 된 것이다. 일반 사람에게 말할 수 있어야 한다. 
# 내가 구현한 것을 보여주는 것이 핵심이다. 
# 데이터를 정제했고 연결한 프로세스를 간략하게 연결하는 것이 핵심이다. 
# 차라리 공식을 대입하자. 그것이 좋을 수 있다. 
# 많이 했지만 와닿지 않다면 그 프로젝트는 실패한 것이다. 
```

```python
# 잘했던 사례는 다음과 같다.
# 대회에 사용했던 PPT를 보면 다음과 같은 특징을 가지고 있다.
# 목차, 선정배경, 문제점확인
# 목표, 하고자 하는 목표
# 진행과정
# 데이터 수집 및 전처리 과정 
# 도식화를 많이 시키는 것이 특징
# 내부 알고리즘을 눈으로 보여주는 방식으로(코드로 보여주지 않고)
# 실제로 사이트를 구현한 이후 메인 기능만 볼 수 있게 해놓자. 
# 최종프로젝트를 할 때에 시연을 할 것이다. 
# 라이브 시연을 많이 하게 된다. 시연의 법칙이 존재하는데 나가기 전에 잘 되어도 나가면 바로 안된다. 
# 그게 시연의 법칙이다. 방금까지 되었지만 갑자기 안되는 경우가 발생한다. 그래서 영상을 많이 쓴다. 
# 시연영상이 뜬다음 가만히 있을 것이다. 영상이 끝나고 다 자고 있다. 영상은 재미없고 화면만 넘어가는 것만 나온다.
# 시연 영상이 나올 때 설명해야 한다. 멈췄다가 해야하는 경우가 많다. 그러면 시간이 너무 지나가버린다. 
# 차라리 더빙을 하자. 안되면 TTS 쓰자. 이렇게 만들어야 사람들이 주목한다. 
# 영상도 깔끔하게 만들면 좋다. 
# 결론이 중요하다. 결론으로 문제점을 어떻게 해소할 수 있는지 말할 수 있어야 한다. 
# 향후 발전 방향도 중요하다. 기간은 정해져있다. 하지만 많은 부분을 수행할 수 있었을 것이다. 
# 즉 많은 생각을 가지고 있었다는 것을 보여줘야 한다. 발전방향을 이야기해야 한다. 
# 그렇게 해야 신경을 많이 쓴 프로젝트라는 느낌을 주게 된다. 
# 그리고 과거의 산출물보다 더 잘 만들어야겠다는 생각이 든다면 성공이다. 
# 당연히 멘토들의 도움이 많이 필요하다. 이러한 부분에서 많이 잡아야 한다. 
```

```python
# 간략화가 핵심이다. 
# 내용을 채우는 것이 핵심이 아니다. 흐름이 핵심이다. 
# 채울 수 있는 것은 다 채우자. 큰 흐름자체는 다 똑같다.
# 모든 사람들은 흐름을 본다. 이상한 흐름으로 가는 경우가 많을 것이다. 
# 사실 네명 다 모를 때나 천재들이 많을 때에 팀이 터진다. 프로젝트는 네 명이서 같이 해봐야 혼자 일하는 경우가 많을 것이다. 
# 내가 할 일을 다 못할 경우 어떻게든 지식을 전수해서라도 할 수 있도록 해야 한다. 그게 역량이고 리더십이다. 
# 특히 기술적으로 잘 하는 사람이 리더가 될 가능성이 매우 높다. 잘 하는 사람들이 리더가 되는 경우가 많다. 
# 네 명이 다 안맞다. 그 프로젝트는 터진 것이다. 
```
