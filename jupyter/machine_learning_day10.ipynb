```python
# 사용자의 취향이나 니즈를 파악하여 적절한 아이템을 추천을 하는 시스템으로 사용자가 상품을 구매할 확률이 올라가게 될 것이다.
# 사용자에게 좀더 친화적인 기업이 될 수 있기 때문에 의미있는 서비스라고 할 수 있다. 
# 본인조차도 몰랐던 취향을 추출해서 서비스를 할 수 있다.
# 고객이 가지고 있는 인사이트를 추출하는 시스템이기도 하다.
# 내가 원하는 영상을 넷플릭스에서 띄우는 것과 매우 유사하다. 
# 이번에 전송하였던 자료는 시퀸셜 리커멘드 시스템이라고 하며 대학교 박사과정에서 수행된 것으로 보여진다. 
# 인공지능을 전공한 연구원들 중에서도 레커멘드 시스템만 연구하는 사람들도 존재한다. 
# 레커멘드 시스템을 이번시간에 확인해 볼 것이다. 
```

```python
# 레커멘드 시스템이란?
# 유튜브 영상의 추천 알고리즘과 같다. 내가 보는 강의와 연관된 강의를 옆에 보여주는 시스템이다. 
# 또는 지금 영상과 이어지는 영상이 유튜브 알고리즘으로 연결된다. 
# 심지어는 영상을 시스템 스스로가 이해를 한다. 눈으로 보고 귀로 듣지 않더라도 이미지, 수치, 기본정보 등으로 연관성을 찾아서 추천해준다.
# 사용자가 관심을 보일 다른 아이템을 연결시켜주는 시스템이다. 
# 추천시스템은 두 부류로 나뉜다. 콘텐츠 기반, 협력 기반으로 나뉜다. 
# 협력기반은 기억기반(KNN기반), 모델기반, 사용자 기반, 아이템기반이 있다.
# 이 모든 것을 합친 하이브리드 기반도 존재한다. 

# 넷플릭스는 협력 필터링을 개발하였다. 모델 기반으로 행렬기반으로 잠재요인을 찾아내는 기법을 사용하였다. 행렬을 분해하여 잠재요인을 추출하는 것이다. 
```

```python
# 콘텐츠 기반
# 2000년대 초반에 만들어진 것이 콘텐츠 기반이었다. 지금은 잘 안쓴다. 상당히 단순한 방식이었다.  
# 사용자가 관심분야에 직접 입력한 정보, 구매내역, 평점을 기반으로 선호 아이템을 파악한 것이다. 
# 선호하는 아이템과 가장 유사한 다른 아이템을 추천한다. 즉 소비자가 선택한 것을 기반으로 추천하는 것이다. 
# 문제는 클레멘타인 같은 영화가 평점이 높다고 해서 진짜로 좋은 영화가 아니었다는 것이다. 
# 또는 내가 산 식품이 맛이 드럽게 없어서 무시했는데 그게 다시 추천을 하는 불상사가 발생하기도 하였다.
# 선택이력이 없는 새로운 아이템을 추천할 수 있는 장점이 있지만 과도한 특수화라는 단점이 존재했고 해당 고객에 대한 추천 성능 보장이 어려웠다.
```

```python
# 협력 필터링, 기억기반
# 특정한 아이템에 대해 선호도가 유사한 고객들은 다른 아이템에도 비슷한 선호를 보인다는 것에서 시작하였다.
# 현재에도 논문으로도 많이 나오고 있는 중이다. 
# 사용자간의 유사도를 추천하여 유사도 높은 이웃이 선택한 아이템 중에서 추천하는 시스템이다. 
# 비슷한 유형의 사람들끼리 크루로 묶은 이후 가장 유사한 사람의 선택을 클릭하는 것이다. 
# 아이템 기반 협력 필터링은 아이템 간 유사도를 측정하여 유사도 높은 아이템을 추천하는 시스템이다. 
# 아이템과 관련된 다양한 정보가 있다면 유사한 사용자를 찾아서 추천해주면 상당히 의미있는 추천시스템이 될 수 있다. 
# 사용자 간의 필터링을 해서 추천하는 시스템으로. 사용자를 기준으로 보는 필터링이다. 비슷한 사용자를 기준으로 추천을 하는 시스템이다.
# 아이템 간의 필터링은 사용자와 반대로 아이템을 기준으로 본다. 아이템은 전체 사용자들의 평점분포를 조사해서 유사한 평점을 찾아서 추천한다.
# 사용자가 비슷한 것을 추천해줄것이냐. 아이템이 비슷한 것을 추천해줄 것이냐에 대한 문제이다. 비슷한 고객을 찾느냐, 비슷한 영화를 찾느냐의 차이이다.  
# 그래서 정보에 따라서 추천시스템이 잘 만들어질 수 있다.  
#----------------------------------------------------------------------------------------------------
# 사용자 기반은 사용자에 대한 정보가 많을 경우에 사용하기 적합하다.
# 아이템 기반은 사용자에 대한 정보가 부족하고 아이템에 대한 정보가 많을 경우에 적합하다. 대체로 아이템이 정확도가 높은 편이다.
# 왜냐하면 아이템쪽의 정보가 훨씬 많은 편이기 때문이다. 
# 다만 다음과 같은 문제점이 발생한다. 
# cold start ploblem : 한 사용자에 대한 충분한 데이터가 부족한 경우 선호도 예측 불가능. (휴면고객 등)
# first rater : 새로운 아이템이 등장하여 평점 점수가 부족한 경우 추천 불가능(신작 등)
# grey sheep problem : 일관성 없는 의견을 가진 사용자들은 데이터 추천에 혼란을 줌(정병이슈 등)
# shilling attack : 악의적으로 평가점수를 긍정, 부정으로 입력하는 경우에 추천에 방해가 됨(클레멘타인 등)
```

```python
# 협력 필터링, 모델 기반
# 데이터에 내제 되어있는 복잡한 패턴을 발견하도록 다양한 모델을 활용한 기법이다.
# 실제 데이터에 적용했을 때 성능이 우수하다. 
# 행렬 분해를 이용해서 추천시스템을 만들려고 한다. 
# 사용자와 아이템 사이에 사용자의 행동과 평점에 영향을 끼치는 잠재적 특성이 있다는 전제를 깔고 시작한다.
# 크기가 크고 복잡한 데이터로도 쉽고 빠르게 분석이 가능하다. 
# 잠재 요인에 따라 행렬분해를 한다. 
# 잠재요인은 무엇인지 모른다. 하지만 영향을 주는 요인이다. 평점을 주는 요인이지만. 말로 설명하기 어렵다. 무의식적인 요소도 존재한다.
# pca랑 상당히 비슷하다. 수 많은 특징, 요인들이 있는데 그 것들이 압축이 된 것으로 볼 수 있다. 
# 잠재요인이 무엇이냐는 답을 못한다. 그냥 잠재 요인인 것이다. 
# 성능에 영향을 주는 부분은 잠재요인의 갯수이다. 다만 이것은 하이퍼퍼러미터이다. 우리가 정해야 한다. 특정한 답이 존재하지 않는다. 
# 반대로 사용자의 잠재 정보행렬과 아이템의 잠재 정보행렬을 곱하여서 숫자를 도출해낸다. 
```

```python
# 딥러닝을 이용한 추천 시스템의 등장
# 딥러닝 기반 추천시스템은 2017년 싱가포르 대학교에서 발표를 하였다. 1000번 이상의 인용이 이루어진 논문이다.
# 모델기반을 사용하는 것의 단점은 새로운 사용자가 등장하거나 기존의 이용자가 사라지는 등의 패턴에 취약하다는 것이다.
# 유사한 사용자를 찾는 법은 jaccar coefficient를 사용한다. 합집합 분의 교집합이다.
# 이것조차 새로운 사용자가 등장시 다른 사용자들과의 유사도를 찾는 것에 한계가 존재했다. 
```

```python
df=chipo[['order_id','item_name']]
df
```

```python
df['order_id']
```

```python
df['order_id'].max()# 주문은 총 1834번 이루어졌다. 
```

```python
df_arr=[[]for i in range(1,df['order_id'].max()+1)] # 효율적이게 살자. 유지보수 귀찮잖아. 일일이 기록하지 말고 컴퓨터한테 이렇게 시키자. 
df_arr # 트랜잭션을 사용하기 위해서는 리스트 안에 저게 다 들어가야 한다. 그래서 이런 식으로 준비한 것이다. 
# 이중 리스트 구조로 1834까지 만드는 것이다. 현재에는 비어있는 리스트라서 아무것도 없다.
# 저걸 채워넣고서야 트랜잭션 인코더(Transaction Encoder)를 사용할 수 있다. 
```

```python
np.shape(df_arr)# 길이는 1834개로 이상없다. 
```

```python
n=0
for i in df['item_name']: # append를 해야한다는 것이 보일 것이다.
    df_arr[df['order_id'][n]-1].append(i)
    n+=1
df_arr
# 복잡하다. 이거 알아두자. 어렵지만 상당히 유용한 코드이다. 
# 1. i는 아이템의 문자이름이다. 즉 아이템의 문자이름을 append한 것이다. 
# 이거 다시 복습하자. 이거 상당히 중요하다. 
```

```python
maxi=0
for i in df_arr:
    if len(i) > maxi:
        maxi = len(i)
print(maxi)
# 주문의 갯수 최대치가 이정도이다. 
# meun=max(len(lst) for lst in df_arr)
# 이것도 꽤 먼치킨스럽네
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te = TransactionEncoder()
te_arr=te.fit(df_arr).transform(df_arr)
te_df = pd.DataFrame(te_arr,columns=te.columns_) # 컬럼스로 하면 이름이 붙을 것이다. 
te_df
# 메뉴 50개에 1834개의 행이 있다. 이제 아페리오리 알고리즘을 적용할 수 있다.
```

```python
apriori(te_df,min_support=0.5) # 지지도가 0.5는 안될것이다 손님 2명중 1명이 특정 아이템을 사야한다는 건데 그건 안될것이다.
```

```python
apriori(te_df,min_support=0.05, use_colnames=True) # 이제야 좀 보인다. 
# 연관규칙을 찾을 수 있는 것이 13부터 슬슬 나온다. 
```

```python
freq_items = apriori(te_df,min_support=0.02, use_colnames=True) # 50명의 사람 중 1명이 같은 것을 주문했다는 이야기이다. 
freq_items # 아이템이 1개인건 의미가 없고 적어도 두 개 이상인 것을 확인해야 할 것이다. 
```

```python
# 아페리오리 사이트 https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/
# 여기서 Example 2 -- Selecting and Filtering Results파트에서 사용할 수 있을 것이다. 
ar_df = association_rules(freq_items,metric='lift',min_threshold=1) # lift가 적어도 1 이상이어야 출력하게 설정하였다. 
ar_df
# 우리가 관심을 가져야 하는 부분은 lift이다. 내림차순 정렬을 해서 배치도 가능할 것이다. 
```

```python
from apyori import apriori
# 재밌는것은 어파이오리에 아페리오리가 들어있다. 
#https://pypi.org/project/apyori/
```

```python
rules = apriori(transactions=df_time['items'].tolist(), min_support=0.002,
                min_confidence=0.02,min_lift=5,min_length=2,max_length=2) # 세부옵션설정이 가능하긴 하다. 서포트만 지정하는 것이 아닌 다른 규칙도 가능하다.  
rules# 아페리오리함수로 만들었다. min,max는 상품의 갯수를 2개씩 제한하는 것이다. 
# 사실상 이 부분이 핵심이다. 
```

```python
res=list(rules)
res# 어질어질하다. 
```
