```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
# 이해를 돕기 위해서 차근차근 설명해보자. 
# c는 color로 색상이다. 
# lw는 line width로 선의 굵기이다. 
# ls는 line style로 라인의 모양이다. --도 되고 -.도 되고 여러 방식으로 가능하다.
# marker 는 데이터의 점 모양이다. o일 수 있고 다른 것일 수 있다.
# ms 는 마커의 크기를 지정한다.
# mec marker edge color로 마커의 테두리 색상이다.  
# mew는 marker edge width로 마커테두리의 굵기이다. 
# mfc는 marker face color이다. 마커 내부의 색을 지정한다.
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
# 범례를 표시한 플롯이다. 현재의 그래프는 코사인 그래프와 사인그래프를 표시한 그래프이다. 
# 이 그래프도 플랏이 각각 두 개가 존재한다.
# legend 에 loc은 location으로 위치를 나타낸다. 1부터 10까지 있다. 디폴트는 최적위치에 알아서 설치한다. 
```

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
# x값을 주고 해당하는 위치에 x라벨에 저장된 이름을 부여한다. 
plt.xticks(x,xlabel)
# x축에 해당하는 바 하나하나에 대한 이름이다. 
# 축에 대한 전체이름을 써 보고 싶다. 축 이름에 종류라고 쓰고 싶다. 
plt.xlabel("종류")
plt.ylabel("빈도수")
# 틱을 부여하고자 한다면
plt.yticks(y)
# y축에 해당하는 숫자가 자동으로 설정되었다. 소숫점 자리가 사라졌다. 
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people)) # 3더하기 10 곱하기(랜덤하게 4개의 난수를 생성)
error = np.random.rand(len(people)) # 0에서 1사이의 임의의 숫자를 입력

plt.title("Barh Chart")# bar horizontal의 약자로 수평바라는 약자이다. 
plt.barh(y_pos, performance, xerr=error, alpha=0.4) # xerr는 신뢰범위를 나타낼 때 사용한다. 알파는 투명도다. 1로 하면 파랑색이된다.
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the brain networks dataset, select subset, and collapse the multi-index
df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)

used_networks = [1, 5, 6, 7, 8, 12, 13, 17]
used_columns = (df.columns
                  .get_level_values("network")
                  .astype(int)
                  .isin(used_networks))
df = df.loc[:, used_columns]

df.columns = df.columns.map("-".join)

# Compute a correlation matrix and convert to long-form
corr_mat = df.corr().stack().reset_index(name="correlation")

# Draw each cell as a scatter point with varying size and color
g = sns.relplot(
    data=corr_mat,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8),
)

# Tweak the figure to finalize
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor(".7")
```
