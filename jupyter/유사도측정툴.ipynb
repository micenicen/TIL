```python
# 코사인유사도 함수
def cosine_yusado(input1,input2):
    sentences = (input1, 
    input2)

    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
    cos_similar = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    return cos_similar[0][0]
cosine_yusado(input1,input2)
```

```python
# 자카드유사도 함수
def Jaccard_yusado(input1, input2):
    answer_string = input1
    input_string = input2

    intersection_cardinality = len(set.intersection(*[set(answer_string), set(input_string)]))
    union_cardinality = len(set.union(*[set(answer_string), set(input_string)]))
    similar = intersection_cardinality / float(union_cardinality)
    return similar
Jaccard_yusado(input1,input2)
```

```python
# SequenceMatcher 함수
def SequenceMatcher_yusado(input1,input2):
    answer_string = input1
    input_string = input2

    answer_bytes = bytes(answer_string, 'utf-8')
    input_bytes = bytes(input_string, 'utf-8')
    answer_bytes_list = list(answer_bytes)
    input_bytes_list = list(input_bytes)

    sm = difflib.SequenceMatcher(None, answer_bytes_list, input_bytes_list)
    similar = sm.ratio()
    return similar
SequenceMatcher_yusado(input1,input2)
```

```python
# 유클리디안 유사도 함수
def Euclidean_yusado(input1,input2):
    sentences = (input1, 
    input2)

    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
    euclidean_dist = euclidean_distances(tfidf_matrix[0:1], tfidf_matrix[1:2])
    return euclidean_dist[0][0]
Euclidean_yusado(input1,input2)
```

```python
# 맨해튼 거리 유사도 함수
def manhattan_yusado(input1,input2):
    # 문장 정의
    sentences = (input1, 
    input2)

    # TF-IDF 벡터화
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)


    # 맨해튼 거리 계산
    manhattan_dist = manhattan_distances(tfidf_matrix[0:1], tfidf_matrix[1:2])
    return manhattan_dist[0][0]
manhattan_yusado(input1,input2)
```

```python
total_score = []
sco = 0
for i in range(len(split_sentences)):
    split_len = len(split_sentences[i])
    sco +=1
    print(sco)
    cosine_default = 0
    Jaccard_default = 0
    SequenceMatcher_default = 0
    Euclidean_default = 0
    manhattan_default = 0
    for j in range(split_len-1):
        input1 = split_sentences[i][j]
        input2 = split_sentences[i][j+1]
        cosine_default += cosine_yusado(input1,input2)
        Jaccard_default +=Jaccard_yusado(input1,input2)
        SequenceMatcher_default +=SequenceMatcher_yusado(input1,input2)
        Euclidean_default += Euclidean_yusado(input1,input2)
        manhattan_default += manhattan_yusado(input1,input2)
    co = cosine_default/(split_len-1)
    Ja = Jaccard_default/(split_len-1)
    Se = SequenceMatcher_default/(split_len-1)
    Eu = Euclidean_default/(split_len-1)
    ma = manhattan_default/(split_len-1)
    score = [co,Ja,Se,Eu,ma]
    total_score.append(score)
```

```python
# 코사인유사도
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sentences = (input1, 
input2)

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
cos_similar = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print("코사인 유사도 측정")
print(cos_similar[0][0])
```

```python
# 자카드유사도
answer_string = input1
input_string = input2

intersection_cardinality = len(set.intersection(*[set(answer_string), set(input_string)]))
union_cardinality = len(set.union(*[set(answer_string), set(input_string)]))
similar = intersection_cardinality / float(union_cardinality)

print(similar)

# 자카드 유사도 (Jaccard Similarity)
# 자카드 유사도는 두 집합 간의 교집합 크기를 합집합 크기로 나눈 값입니다. 이 방법은 두 문장에서 공통적으로 등장하는 단어의 비율을 측정합니다.
# 이 방식은 특히 짧은 텍스트나 키워드 중심의 텍스트 분석에 유용합니다.
```

```python
# SequenceMatcher 이용
import difflib

answer_string = input1
input_string = input2

answer_bytes = bytes(answer_string, 'utf-8')
input_bytes = bytes(input_string, 'utf-8')
answer_bytes_list = list(answer_bytes)
input_bytes_list = list(input_bytes)

sm = difflib.SequenceMatcher(None, answer_bytes_list, input_bytes_list)
similar = sm.ratio()

print(similar)

# SequenceMatcher 유사도
# difflib의 SequenceMatcher는 두 문자열 간의 유사도를 측정하는 데 사용되며, 특히 문자 단위에서의 유사성을 평가합니다.
# 이 방법은 두 문자열이 얼마나 많은 부분을 공유하는지를 측정하며, 길이가 다른 문자열 간의 비교에 적합합니다.
```
