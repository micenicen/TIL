```python
# 내장함수와 외부함수
# 따로 쓰겠다고 명시하지 않아도 사용 가능한 함수이다. 일반적인 프로그램을 하는 과정에서 많이 사용되는 특징이 있다.
# 예를들어 print, input함수 같은 것들이 존재한다.
# random, pickle같은 모듈이 존재하고 모듈을 묶어놓은 패키지가 존재한다. 이것들은 불러오지 않으면 사용할 수 없다.
# import 명령은 내장되어있지 않고 불러와야만 사용할 수 있다.
# 일반적으로 'import 모듈' 이런식으로 불러오거나  import 모듈1, 모듈2, 모듈3 ... 같이 여러 모듈을 불러올 수 있다.
# 모듈을 가져온 목적은 모듈 내부의 변수와 함수를 사용하기 위함이다. 모듈 다음에 점을 찍고 변수, 함수를 기재해서 사용할 수 있게 된다.
import math
print(math.sqrt(4))# 예를들면 수학모듈을 불러와서 제곱근을 구하는 방식은 이와같이 이루어진다.
# 모듈을 가져오고 모듈.함수와 같은 방식으로 모듈을 가동시켰다. 이러한 방식이 일반적이다.
```

```python
#정규표현식
# 원본데이터로부터 불필요한 자료를 제외하고 필요한 자료만 추출하는 과정에서 사용된다. 
# 복잡한 문자열을 처리할 때 사용하는 기법이다. 불필요한 문자열을 제외하고 사용하고자하는 문자열만 추출할 수 있다.
# 파이썬 뿐만 아니라 다른 프로그래밍 언어에서도 매우 많이 사용된다. 다만 많은 사람들이 정규 표현식을 모르는 사람들이 많다.
# 필요시 정규표현식을 사용하는 것이다. 필수가 아닌 옵션인 것이다. 
# 어려운 것이 단점이다. 정규표현식을 사용하기 위해서는 정규표현식의 기호를 알아야 한다.
# https://wikidocs.net/4308에서 확인해보자.
```

```python
import re #정규표현식 모듈
print(re.match("hi",'hihello'))# 일치여부 확인 -(패턴, 문자열) 문자열에 패턴이 존재합니까?(반대로 읽는 것 주의)
print(re.match("bye",'hihello'))
# <re.Match object; span=(0, 2), match='hi'> 매치된 오브젝트의 위치는 0부터 1까지 매치되었다.hi가 매치되었다.
print(re.match("hi",'ahihello')) #첫번째 문자가 매치가 안되면 뒤에 매치가 되든 말든 검사하지 않는다.
print(re.match("hi",'haihello'))# 앞에있는게 매치되어도 뒤의 다른게 매치 안되면 매치가 아니라고 한다.
print(re.search("hi",'ahihello'))# 첫문자가 매치되지 않아도 매치가 된 것으로 나온다.
print(re.search("hi",'aihellohi'))# 맨 뒤의 문자는 매치가 되었다. 그렇기에 매치가 된 것으로 나온다. 
#즉 match는 처음부터 나왔냐.  search는 전체검새를 한다는 느낌이다. 
print(re.search("hi",'ahihellohi'))#왼쪽부터 읽어서 처음나온 것을 결과로 내놓고 종료한다. 즉 1개만 찾는다.
print(re.findall("hi",'ahihellohi'))#두개의 존재를 확인할 수 있다. 모든 결과를 찾는다. 그렇기에 많이 사용된다.
```

```python
# 정규표현식 문자클래스
# 대괄호는 문자클래스이다. 어느 문자든 다 올 수 있다.
#[abc]는 a 또는 b 또는 c 중 매치가 되는가? 이다. 
print(re.search("[abc]", "a"))#대괄호에는 반드시 겹따옴표를 사용해야 한다.
print(re.search("[abc]", "hihello"))
print(re.match("[abc]", "hi ani")) # 앞 글자에 바라 안나오면 매치는 바로 끝난다. 
print(re.match("[abc]", "before")) 
print(re.findall("[a-z]", "before")) # 모든 알파벳을 포함한다면 다음과 같이 할 수 있다. 
print(re.findall("[a-zA-Z]", "Before")) # 대문자도 이렇게 할 수 있다. 모든 알파벳을 포함한다. 
print(re.findall("[a-zA-Z0-9]", "Before123")) # 숫자도 다음과 같이 표시할 수 있다.
print(re.findall("[ㄱ-ㅎ가-힣]", "마라탕배달꾼")) # 한글은 다음과 같이 불러오는 것이 가능하다. 
```

```python
wf = """
[앵커]

폭염주의보가 다시 전국 곳곳으로 확대됐습니다.

당분간 한낮에 대부분 지역의 기온이 30도를 웃돌아 덥겠습니다.

[리포트]

충청과 남부지방에 이어 수도권과 강원도 홍천에도 폭염주의보가 내려졌습니다.

다시 전국 곳곳으로 폭염주의보가 확대됐는데요.

오늘 서울과 광주, 대구의 낮 최고 기온은 32도까지 오르겠습니다.

내일과 모레도 서울의 낮 기온 31도 등 당분간 한낮엔 30도 안팎까지 올라가는 더위가 이어지겠습니다.

하지만 아침 기온은 점차 낮아지겠습니다.

오늘 아침엔 서울 25도 등 지난 밤에 열대야가 나타났지만 내일 아침엔 23도, 모레 19도 등 아침 기온이 20도 안팎까지 내려가서 아침, 저녁엔 비교적 선선해지겠습니다

오늘 오후에도 소나기가 올 가능성이 있습니다.

5에서 최대 40밀리미터의 소나기가 오는 곳이 있겠고, 국지적으로 천둥, 번개가 치면서 시간당 30밀리미터 안팎의 강한 소나기가 오는 곳이 있을 것으로 예상됩니다.

기상정보였습니다.

노은지 기상캐스터/그래픽:강은지
"""
print(re.findall("[ㄱ-ㅎ가-힣0-9]", wf))
```

```python
# 정규표현식을 쓰는 하나의 방법이다. 웹 크롤링을 한 결과에 대해서 원하는 자료를 추출하고자 할때 사용한다.
# 꺾새(^)는 not의 의미이다. 즉 꺾새를 붙이면 그 자료를 제외하고 찾는다. 
print(re.findall("[^ㄱ-ㅎ가-힣0-9 \n.,]", wf)) #맨 앞에 붙일 경우 전체부정을 나타낸다.
# 주의할 점은 대괄호 안에 있는지 밖에 있는지 여부이다. 아예 뜻이 달라진다. 
# \d = [0-9]
print(re.findall("\d", wf)) #바로 사용할 경우 이런식으로 역슬래시 알파벳을 사용할 수 있다.
# \D = [^0-9]
print(re.findall("\D", wf))
```

```python
# \s = [ \t\n\r\f\v] 줄바꿈, 공백 문자 등 보이지않는 문자들을 말한다.
print(re.findall("\s", wf))
# \S = [^ \t\n\r\f\v] 줄바꿈, 공백 문자 등 보이지않는 문자들을 제외한다.
print(re.findall("\S", wf))
```

```python
# \w = [a-zA-Z0-9_] 문자와 숫자들을 말한다. 신기한 점은 모든 문자로 한글도 포함한다.
print(re.findall("\w", wf))
# \W = [a-zA-Z0-9_] 문자와 숫자들을 제외한다.
print(re.findall("\W", wf))
```

```python
print(re.match("[1-5]","12345"))#대괄호를 씌워서 범위탐색을 하는 것이 가능하다.
print(re.match("[1-5]+","12345"))#패턴에 부합되는 문자가 하나 이상이 있는지 확인한다.
print(re.match("[1-5]+","612345"))#물론 기능 자체가 맨 앞에 찾는게 없으면 종료시킨다.
print(re.match("[1-5]+","341612345"))#이러면 6 이전까지 매치결과를 찾는다.
```

```python
print(re.match("[1-5]*","341612345"))#패턴에 부합되는 문자가 0개 이상 있는지 확인한다. 즉 없어도 매치된다.
print(re.match("a*" , "123"))
```

```python
#비교하기
print(re.match("a" , "123")) # 매치는 없으면 그냥 작동하지 않는다.
print(re.match("a" , "a123")) # 매치가 될 것이다.
print(re.match("a" , "aaaa123")) # 단 한개만 매치될 것이다.
print(re.match("a+" , "aaaa123")) # 4개의 a가 매치될 것이다.
print(re.match("a*" , "123")) # 출력이 되긴 했으나 빈칸이 매치되었다.
print(re.match("a*b+" , "baaa123"))# a는 0개인데 b가 매치되었다.
print(re.match("ba*[1-3]" , "baaa123")) #1부터 3까지지만 플러스가 없어서 1까지 나올 것이다.
print(re.match("ba*[1-3]+" , "baaa123")) #전체 다 출력될 것이다.
#즉 + 나 * 가 안 붙으면 1개만 본다. +를 보면 해당되는게 나오면 안나올때까지 본다. 별을 붙이면 0개 이상 나오는 것이기에 매치된다. 
#그렇기에 연습이 필요하다. 
```

```python
print(re.match("[0-9a-z]*" , "Ahi1234")) #여기서 매치는 0개가 될 것이다.
print(re.match("[0-9a-z]*가" , "hi1234가나다")) #문자 끝나고 가가 존재하기에 '가'까지 매치된다.
print(re.match("[0-9a-z]+가" , "hi1234가나다")) #문자 끝나고 가가 존재하기에 '가'까지 매치된다.
print(re.match("[0-9a-z가-힣]*" , "hi1234가나다")) #전부 매치될 것이다.
print(re.match("[0-9a-z가-힣ㅋ-ㅎ]*" , "hi1234가나다ㅋㅋㅎ")) #전부 매치될 것이다.
print(re.match("a*b" , "b")) #a가 0개 충족되어서 다음으로 넘어가고 b가 매치될 것이다.
print(re.match("a*b" , "aab")) # 전부 매치될 것이다.
```

```python
# 점은 엔터 제외 모든 문자를 포함한다.
print(re.match("a.b" , "aab")) # 가운데에 아무 문자나 와도 되기에 매치된다.
print(re.match("a.b" , "abc")) # a빈칸b까지는 만족하지만 c를 만족하지 않기에 매치되지 않는다.
print(re.match("a.b" , "a$b")) # 매치된다. 특수문자도 해당된다.
```

```python
#파일명과 확장자가 올바른지 알아보기 위해서는 점을 확인해야 한다. 이럴때는 다음과 같이 작성하자
res = re.match("[a-zA-Z0-9]+[.][a-zA-z]" , "test2")
if res: 
    print("정상적인 파일입니다.")
else:
    print("비정상적인 파일입니다.")
```

```python
# ?문자
# ? 앞에 문자 또는 범위가 0개인지 1개인지 확인하는 용도로 사용한다.
print(re.match("abc?d","abd")) # 0개가 나왔으니 매치가 되었다.
print(re.match("ab[0-9]?d","abd")) # 매치가 되었다.
print(re.match("ab[0-9]?d","ab7d")) # 매치가 되었다.
print(re.match("ab[0-9]?d","ab78d"))# 매치가 되지 않는다. 8이 아니라 d가 바로나오지 않았기 때문이다.
```

```python
# 중괄호문자
print(re.match("h","hhheee")) #기본형
print(re.match("h{3}","hhheee"))# 3개가 매치가 되었다.
print(re.match("h{3}","hhhheee"))# 중괄호의 횟수만큼만 매치되었다.
print(re.match("h{5}","hhheee"))#중괄호의 횟수만큼 매치되지 않으면 매치되지 않는다.
print(re.match("h{3}e","hhheee"))# 매치되었다. 중괄호만큼 반복하고 바로 e가 나왔다.
print(re.match("h{3}e{3}","hhheee"))# 전체 매치되었다.
```

```python
print(re.match("hi","hihi"))
print(re.match("(hi)","hihi"))#사실 위의것과 같다. 그러나 괄호를 친 이유는 hi를 하나로 묶기 위해서이다.
print(re.match("(hi){2}","hihi"))#전체 매치가 되었다.
print(re.match("(hi){3}","hihi"))#3번반복이지만 주어진 것은 2번 반복이라 매치되지 않는다.
print(re.match("(hi){3}h","hihihihi"))#중괄호안에 반복된 수만큼 충족하고 다음의 h도 충족하였다.
print(re.match("(hi){3}[0-9]","hihihi"))# 곱 기호가 안붙으면 반드시 나와야 한다. 나오지 않으면 매치되지 않는다.
print(re.match("(hi){3}[0-9]*","hihihi"))# 별은 존재하지 않아도 0개를 출력하기에 매치된다.
```

```python
print(re.match("[0-9]{3}","123")) # 될 것이다. 0부터 9까지 3번 반복되야하는데 3개 나왔다.
print(re.match("[0-9]{3}","12")) #3개가 반복되ㅣ어야 하지만 2개만 반복되어서 출력되지 않았다.
print(re.match("[0-9]{2,3}","12")) #2개나 3개나 전부 허용하기에 이상없이 출력된다.
print(re.match("[0-9]{2,6}","12345")) #2개가 충족되는 것이 전부기에 전부 해당된다.
print(re.match("[0-9]{2,6}","1")) #2개조차도 충족하지 못한다. 매치되지 않는다.
```

```python
#전화번호 검증 패턴식 정의하기
#1) 지역번호- 국번호- 전화번호 순으로 작성되야 한다.
#2) 2 or3 - 3 or 4 - 4 순으로 작성된다.
#3) 모두 숫자로 작성되어야 한다.
num = input()
number1 = re.match("[0-9]{2,3}-[0-9]{3,4}-[0-9]{4}",num)
if number1:
    print("이상없음")
else:
    print("불가")
```

```python
#컴파일
p = re.compile('ab*') # 패턴을 컴파일로 따로 정의했다.
print(re.match(p,"ab")) # 그리고 기존의 방법과 같이 매치여부를 확인할 수 있다.
# 연습은 매치로 했지만 실질적으로 findall이 사용빈도가 높다.
```

```python
print(re.match("[가-힣]","홍길동")) #한개만 매치
print(re.match("[가-힣]*","홍길동")) #전부 매치
print(re.match("[가-힣]+","홍길동")) #전부 매치
print(re.match("[A-Z]","Hello")) #한개만 매치
print(re.match("[^A-Z]","Hello")) #시작이 대문자가 아니어야 하는데 대문자라서 매치안됨
print(re.match("^[A-Z]","Hello")) #시작문자가 대문자인가? 에서 해당되어서 매치됨


```

```python
print(re.search("*","2 ** 3")) # 별이 있는지 없는지 확인하고자 한다. 하지만 별은 별 앞에 있는 대상에 대해서 0번 이상 나온다는 의미이다.
# 하지만 반복적으로 나올 것이 아무것도 나오지 않아서 애러가 출력된다.
```

```python
print(re.search("\*","2 ** 3")) #역슬래쉬는 별 자체를 출력하게 한다.
print(re.search("\*+","2 ** 3")) #별 두개 이후도 출력하는 것이니 별 두개가 매치될 것이다. 
print(re.search("[*]+","2 ** 3")) #아예 귀찮으면 대괄호로 묶어버리면 그만이다.
```

```python
print(re.match("[a-zA-Z0-9()$]+","$(Doc9)")) # 특수문자, 대소문자, 숫자까지 전부 정의하는 것을 만들때 단순하게 만들면 이렇게 가능하다.
```

```python
print(re.match("[0-9]+ [0-9]+","123 45")) # 공백으로 구분이 되어져 있다. 이것을 일반적으로 출력하는 것은 다음과 같다.
res = re.match("([0-9]+) ([0-9]+)","123 45") #앞뒤의 자료를 따로 분리하고 싶을 경우가 있을 것이다. 소괄호를 그룹으로 배정한다.
print(res.group())# 매치된 것 전체가 res에 저장이 되어진 상태이다.
print(res.group(1)) # 소괄호를 묶음으로서 소괄호 단위로 그룹으로 간주된다.
print(res.group(2)) # 패턴에 부합되는 경우에 패턴식을 만족하는 그룹식을 그룹단위로 묶어서 확인할 수 있다.
print(res.groups()) #그룹단위를 확인하는 것으로 튜플단위로 나온다.
```

```python
#문자열이 주어진 상황에서 함수명과 인수를 각각의 그룹으로 나눠서 출력하고 싶다고 하자. 그룹0에 출력을 하면 print가 나오게 하고 1은 123이 나오게 하고 싶다 하자.
res = re.match("([a-z]+)\(([0-9]+)\)","print(123)")
res.group(0)
# 그룹 0은 전체출력이다....어떻게 해야할까?
```

```python
res = re.match("([a-zA-Z][a-zA-Z_0-9]+)","print(123)") # 이는 첫 문자는 영단어 또는 언더바로만 시작해야한다는 의미이다. 
```

```python
#그룹에는 이름을 부여할 수 있다. 그룹 맨 앞에다가 이름을 붙일 수 있는데.  
res5 = re.match("(?P<fn>[a-zA-Z][a-zA-Z_0-9]+)\((?P<arg>[0-9]+)\)","print(123)")
print(res5.group("fn"))
print(res5.group("arg"))
```
