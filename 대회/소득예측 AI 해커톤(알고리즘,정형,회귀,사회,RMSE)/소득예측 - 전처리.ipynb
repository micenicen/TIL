```python
# 소득 훈련파일 호출
import numpy as np
import pandas as pd

soduk = pd.read_csv('train.csv')
soduk   
```

```python
soduk.info()
# 1. 아이디는 드랍한다. 단지 순서만을 나타내며 소득과는 연관이 없다. 
# 2. 나이는 남겨둘 것이다. 이것은 매우 중요한 정보이다. 하지만 카테고리화가 필요할
# 것이다. 여기서 아이(~18), 청년(18-35), 중년(36-65), 노년 순으로 정리하자. 
# 3. 성별 또한 카테고리가 적으면서 매우 중요하다. 
# 4. 학력은 카테고리화가 반드시 필요하다. 학교 카테고리화는 반드시 진행할 것이다. 
# 5. 고용상태에 따른 분류도 매우 중요하다. 그대로 사용한다.
# 6. 주 노동량은 53개의 카테고리로 존재한다. 4 단위로 카테고리화 시킬 예정이다.
# 7. 산업고용분포의 4688과 직업상태의 4688은 동일하다. 여기에 주목하자. Unknown을 
# 먼저 Not in universe or children 로 대체하자. 어쩌면 직업상태가 카테고리가 더 적어서
# 쓸만해보인다.
# 8. 인종은 매우 중요한 지표다. 
# 9. 히스패닉여부는 좀 더 지켜보자. 미국에는 히스패닉이 많기에 어디출신인지까지
# 신경쓸지는 미지수이다. 
# 10. 결혼은 매우 주요하다. 
# 11. 가구구성은 간략화 구성을 사용한다. 우리는 마약탐지를 하는게 아니다. 
# 12. 시민권은 당연히 중요하다. 
# 13. 출신국가는 대륙별로 분류하자. 너무 복잡해진다.
# 14. 세금보고서도 중요하다.
# 15. Gains, Losses도 중요성이 있을지라도 다른 요소들보다 중요하다고 생각되지 않는다.
# 16. 배당이 소득과 상관없다고 해도 연관성분석은 필요할 것이다. 
# 17. 소득수준은 중요하다.
# 18. 답안지인 소득은 최종제출은 5단위로 분류해서 제출하도록 계획하자
```

```python
# 1. ID열 제거
soduk.drop('ID', axis=1 ,inplace = True)
```

```python
# 2. 나이 카테고리화
# 나는 나이를 4 단위로 분류할 것이다.
soduk['Age']
```

```python
def classify_age(age):
    if age <= 17:
        return 'Child'
    elif 18 <= age <= 35:
        return 'Youth'
    elif 36 <= age <= 65:
        return 'Middle'
    else:
        return 'Silver'
soduk['Age_Category'] = soduk['Age'].apply(classify_age)

soduk['Age_Category']
    
```

```python
# 3. 성별은 그냥 둔다. 그대로도 원핫 인코딩에 활용 가능하다. 
# 4. 학력은 축약한다. 옆에 있는 숫자는 걸리적거린다. 
soduk['Education_Status']
```

```python
def Education(edu):
    if  edu == "High Freshman" or edu == "High Junior" or edu == "High Sophomore" or edu == "High Senior":
        return 'High'
    elif edu == "Elementary (5-6)" or edu == "Elementary (1-4)":
        return 'Elementary'
    elif edu == 'Middle (7-8)':
        return 'Middle'
    else:
        return edu
soduk['Education_Category'] = soduk['Education_Status'].apply(Education)

soduk['Education_Category']
```

```python
soduk['Education_Category'].value_counts()
```

```python
def classify_working_week(weeks):
    if weeks == 0:
        return 'Not_Working'
    elif 1 <= weeks <= 13:
        return 'Part-Time'
    elif 14 <= weeks <= 26:
        return 'Semi-Full-Time'
    else:
        return 'Full-Time'

soduk['Working_Category'] = soduk['Working_Week (Yearly)'].apply(classify_working_week)
soduk['Working_Category']
# 시간이 0부터 52까지 너무 많다. 이걸 4개의 카테고리로 나눠보자. 
```

```python
# 먼저 스택을 이용해서 Not in universe or children이 정마로 unknown과 일치하는지 확인하자. 
# 우연의 일치일 가능성도 있기 때문이다.
stack = 0
search = soduk['Industry_Status'] 
find = soduk['Occupation_Status'] 
for i in range(len(search)):
    if search[i] == 'Not in universe or children':
        if find[i] =='Unknown':
            stack+=1
print(stack)
# 결과값은 4688이다. 즉 Not in universe or children과 Unknown은 일치하는 항목이다. 
```

```python
# 인종은 그대로 사용한다. 
# 히스패닉은 이번에 사용하지 않는다. 추후 사용할 경우 Hispanic으로 일괄 전환시킨 다음 인종에 끼워넣을 것이다. 
```

```python
# 결혼 자료는 그대로 쓸 것이다. 가구구성 간략화만 사용할 것이다. 
# 시민권은 그냥 둘 것이다. 그 자체로 유용하다.
```

```python
# 출생지는 중요하다. 미국과 대륙을 나눠야 한다. 
def classify_country(country):
    if country == 'US' or country == 'Outlying-U S (Guam USVI etc)':
        return 'US'
    elif country in ['Mexico', 'Puerto-Rico', 'Cuba', 'Dominican-Republic', 'El-Salvador', 'Guatemala', 'Honduras', 'Nicaragua', 'Panama']:
        return 'North_America'
    elif country in ['Columbia', 'Ecuador', 'Peru']:
        return 'South_America'
    elif country in ['Philippines', 'India', 'South Korea', 'Vietnam', 'Japan', 'China', 'Taiwan', 'Hong Kong', 'Cambodia', 'Laos', 'Thailand']:
        return 'Asia'
    elif country in ['Germany', 'Canada', 'England', 'Poland', 'Italy', 'Portugal', 'Ireland', 'France', 'Greece', 'Scotland', 'Yugoslavia', 'Hungary', 'Holand-Netherlands']:
        return 'Europe'
    else:
        return 'Unknown'

# 새로운 카테고리 열 추가
soduk['Birth_Country_Category'] = soduk['Birth_Country'].apply(classify_country)
# 아프리카와 오세아니아 출신 사람들은 아마 Unknown에 있을 것이다. 
```

```python
# 세금보고서는 그대로 쓸 것이야. 그리고 Gains와 Losses는 쓰지 않을것이다. 
# 분석에서 제외될 변수: "ID", "Gains", "Losses", "Dividends"
# 분석에 포함될 주요 변수: "Age", "Gender", "Education_Status", 
# "Employment_Status", "Working_Week (Yearly)", "Race", "Martial_Status", 
# "Household_Summary", "Citizenship", "Birth_Country" (대륙별로 분류), 
# "Tax_Status", "Income_Status"
```

```python
soduk.info()
```

```python
# 다음과 같이 가공하자.
# 'Age_Category' : 원핫 인코딩으로 가공
# "Gender" : 원핫 인코딩으로 가공
# 'Education_Category' : 원핫 인코딩으로 가공
# 'Working_Category' : 원핫 인코딩으로 가공
# 'Occupation_Status' : 원핫 인코딩으로 가공
# "Race" : 원핫 인코딩으로 가공
# "Martial_Status" : 원핫 인코딩으로 가공
# "Household_Summary" : 원핫 인코딩
# "Citizenship" : 원핫 인코딩
# "Birth_Country_Category" : 원핫 인코딩
# "Tax_Status" : 원핫 인코딩
# "Income_Status" : 원핫 인코딩
# 원-핫 인코딩을 적용할 열만 선택
columns_to_encode = [
    'Age_Category', 'Gender', 'Education_Category', 'Working_Category', 
    'Occupation_Status', 'Race', 'Martial_Status', 'Household_Summary', 
    'Citizenship', 'Birth_Country_Category', 'Tax_Status', 'Income_Status'
]

# 지정된 열에 대해서만 원-핫 인코딩 적용
soduk_encoded = pd.get_dummies(soduk, columns=columns_to_encode)
```

```python
soduk_encoded.info()
# 여기서 필요없는 열을 드랍하자. 
```

```python
# 드랍열 설정
columns_to_drop = [
    "Age", "Education_Status", "Employment_Status", "Working_Week (Yearly)", 
    "Industry_Status", "Hispanic_Origin", "Household_Status", 
    "Birth_Country", "Birth_Country (Father)", "Birth_Country (Mother)", 
    "Gains", "Losses", "Dividends"
]

# 지정된 열들을 드랍
soduk_encoded_dropped = soduk_encoded.drop(columns=columns_to_drop)

# 드랍된 이후의 데이터프레임 확인
soduk_encoded_dropped

# 이렇게 전처리를 완료하게 되었다. income부분은 답지에 해당된다. 
# 답지를 제외하면 모두 
```

```python
# 이제 훈련데이터와 테스트 데이터를 나눠보자.
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(        # 트레인데이터, 테스트데이터 분리결과
    soduk_encoded_dropped.drop(['Income'],axis=1),soduk_encoded_dropped.Income,        # 정답 데이터 드롭한것이 트레인, 정답 데이터는 답지
    test_size=0.2,random_state= 41)                   # 테스트용으로는 20%만 사용, 랜덤값은 41로 고정
```

```python
# 로지스틱으로 해보자. 
from sklearn.linear_model import LogisticRegression # 사이킷런 리니어모델 - 로지스틱 회귀
clf = LogisticRegression()                          # 로지스틱회귀모델 제작
clf.fit(xtrain,ytrain)                              # 모델링 절차 - 회귀모델 제작
ans = clf.predict(xtest)                                  # 테스트 문제풀이
ans
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 성능 평가
print("Accuracy:", accuracy_score(ytest, ans))
print("Confusion Matrix:\n", confusion_matrix(ytest, ans))
print("Classification Report:\n", classification_report(ytest, ans))
# 정확도가 고작 42%밖에 안된다. 어쩔 수 없다. 소득은 범위기 때문에 정확하게 보는건 어렵다. 
```

```python
# 선형회귀는 다음과 같이 진행한다. 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 특성과 타겟 변수를 준비합니다.
X = soduk_encoded_dropped.drop(['Income'], axis=1) # 타겟 변수를 제외한 특성만 사용
y = soduk_encoded_dropped['Income'] # 타겟 변수

# 데이터를 훈련 세트와 테스트 세트로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)

# 선형 회귀 모델 객체를 생성합니다.
model = LinearRegression()

# 훈련 데이터를 사용하여 모델을 학습시킵니다.
model.fit(X_train, y_train)

# 테스트 데이터에 대한 예측을 수행합니다.
y_pred = model.predict(X_test)

# 성능 평가: 여기서는 평균 제곱 오차(MSE)와 결정 계수(R^2)를 사용합니다.
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("평균 제곱 오차(MSE):", mse)
print("결정 계수(R^2):", r2)
```

```python
# 랜덤 포레스트로 확인하기
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 랜덤 포레스트 회귀 모델 객체를 생성합니다.
# n_estimators는 결정 트리의 수를 의미합니다. 여기서는 예시로 100개의 트리를 사용합니다.
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# 훈련 데이터를 사용하여 모델을 학습시킵니다.
rf.fit(X_train, y_train)

# 테스트 데이터에 대한 예측을 수행합니다.
y_pred_rf = rf.predict(X_test)

# 성능 평가
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("랜덤 포레스트 회귀의 평균 제곱 오차(MSE):", mse_rf)
print("랜덤 포레스트 회귀의 결정 계수(R^2):", r2_rf)

# 결정계수가 22에서 13으로 하락했다. 아까보다 더 안좋다.
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 그래디언트 부스팅 회귀 모델 객체를 생성합니다.
gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,
                                max_depth=3, random_state=42)

# 훈련 데이터를 사용하여 모델을 학습시킵니다.
gbr.fit(X_train, y_train)

# 테스트 데이터에 대한 예측을 수행합니다.
y_pred_gbr = gbr.predict(X_test)

# 성능 평가
mse_gbr = mean_squared_error(y_test, y_pred_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)

print("그래디언트 부스팅 회귀의 평균 제곱 오차(MSE):", mse_gbr)
print("그래디언트 부스팅 회귀의 결정 계수(R^2):", r2_gbr)
# 좀더 나아졌다. 24정도면 괜찮아졌다. 
```
