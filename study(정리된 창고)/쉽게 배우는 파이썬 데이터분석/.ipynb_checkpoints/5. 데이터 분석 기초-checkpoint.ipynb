```python
df = pd.DataFrame({'var1' : [4,3,8],
                   'var2' : [2,6,1]})
df
```

```python
df['var_sum'] = df['var1'] + df['var2']          # 더한값 변수 만들기
df
```

```python
df['var_mean'] = (df['var1'] + df['var2'])/2    # 더한값 평균 만들기 mean함수 사용
df
```

```python
import numpy as np
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
mpg['total'] = (mpg['cty']+mpg['hwy'])/2           # 연비 작성
mpg['test']= np.where(mpg['total'] >= 20, 'pass','fail' ) # 테스트값 사용. np.where은 if문처럼 사용
mpg.head()
```

```python
mpg['test'].value_counts()
```

```python
count_test = mpg['test'].value_counts()
count_test.plot.bar(rot=0)                     #축 이름 수평으로 맞추기
```
