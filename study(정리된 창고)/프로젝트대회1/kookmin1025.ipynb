```python
# text_keyword결측값 대체하기
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from sklearn.preprocessing import OneHotEncoder

apply_train=pd.read_csv('apply_train.csv')
# 지원한 회사를 one-hot encoding으로 표현

# 지원한 회사를 one-hot encoding으로 표현
ohe = OneHotEncoder()
encoded = ohe.fit_transform(apply_train[['recruitment_seq']])

# 다음 줄을 수정합니다.
try:
    column_names = ohe.get_feature_names_out(['recruitment_seq'])
except AttributeError:
    column_names = ohe.get_feature_names(['recruitment_seq'])

encoded_df = pd.DataFrame(encoded.toarray(), columns=column_names)
apply_train_ohe = pd.concat([apply_train[['resume_seq']], encoded_df], axis=1).groupby('resume_seq').sum().reset_index()

similarity = cosine_similarity(apply_train_ohe.iloc[:, 1:])

# 유사도 행렬을 DataFrame으로 변환
similarity_df = pd.DataFrame(similarity, index=apply_train_ohe['resume_seq'], columns=apply_train_ohe['resume_seq'])

# 결측값 대체 함수
def fill_missing_keywords(row):
    if pd.isnull(row['text_keyword']):
        # 유사도가 가장 높은 지원자들을 정렬
        sorted_similar_resumes = similarity_df[row['resume_seq']].sort_values(ascending=False).index.tolist()
        
        # 유사도 순으로 NaN이 아닌 text_keyword 값을 찾기
        for similar_resume in sorted_similar_resumes:
            value = data[data['resume_seq'] == similar_resume]['text_keyword'].values[0]
            if not pd.isnull(value):
                return value
    return row['text_keyword']

# 결측값 대체
data['text_keyword'] = data.apply(fill_missing_keywords, axis=1)
data
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def build_user_company_matrix(users, companies, applications):
    # Initialize matrix with zeros
    matrix = np.zeros((len(users), len(companies)))
    
    # Set values to 1 based on applications
    for _, row in applications.iterrows():
        user_idx = users.index.get_loc(row['resume_seq'])
        company_idx = companies.index.get_loc(row['recruitment_seq'])
        matrix[user_idx, company_idx] = 1

    return matrix

# 데이터 로드
application_data = pd.read_csv('apply_train.csv')

# 사용자-회사 지원 매트릭스 생성
user_company_matrix = build_user_company_matrix(user_features, company_features, application_data)

# 사용자와 회사 간의 유사도 계산 (여기서는 사용자-회사 지원 매트릭스를 사용)
user_to_company_similarity_matrix = cosine_similarity(user_features, user_company_matrix)

def get_all_recommendations(user_to_company_similarity_matrix, applied_jobs):
    all_recommendations = {}
    
    all_users = user_features.index.tolist()
    for user_id in all_users:
        recommended_jobs = recommend_jobs_based_on_similarity(user_id, user_to_company_similarity_matrix, applied_jobs)
        all_recommendations[user_id] = recommended_jobs
    
    return all_recommendations

all_user_recommendations = get_all_recommendations(user_to_company_similarity_matrix, application_data)

# 결과값을 DataFrame으로 변환하고 CSV 파일로 저장
recommendations_df = pd.DataFrame(all_user_recommendations, columns=['resume_seq', 'recommended_job_id'])
recommendations_df.to_csv('user_recommendations.csv', index=False, encoding='utf-8-sig')
```

```python
# 앙상블기법이 적용된 코드
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def compute_similarity(data_features):
    similarities = cosine_similarity(data_features)
    return pd.DataFrame(similarities, index=data_features.index, columns=data_features.index)

def recommend_jobs_based_on_similarity(user_id, similarity_matrix, application_data, top_k=5):
    similar_items = similarity_matrix.loc[user_id].sort_values(ascending=False).head(top_k + 1).index[1:]
    applied_jobs = set(application_data[application_data['resume_seq'] == user_id]['job_id'])
    recommended_jobs = [job for job in similar_items if job not in applied_jobs]
    return recommended_jobs[:top_k]

def ensemble_recommendations(user_based_recs, job_based_recs):
    # 여기서는 간단히 두 리스트를 합친 후 중복을 제거합니다.
    # 다양한 결합 방법을 적용할 수 있습니다.
    return list(set(user_based_recs + job_based_recs))

user_features = pd.read_csv('user_features.csv').set_index('resume_seq')
job_features = pd.read_csv('job_features.csv').set_index('job_id')
application_data = pd.read_csv('apply_train.csv')

user_similarity_matrix = compute_similarity(user_features)
job_similarity_matrix = compute_similarity(job_features)

all_recommendations = {}

all_users = application_data['resume_seq'].unique()
for user_id in all_users:
    user_based_recs = recommend_jobs_based_on_similarity(user_id, user_similarity_matrix, application_data)
    job_based_recs = recommend_jobs_based_on_similarity(user_id, job_similarity_matrix, application_data)
    
    final_recommendations = ensemble_recommendations(user_based_recs, job_based_recs)
    all_recommendations[user_id] = final_recommendations

# 결과를 DataFrame 형태로 변환
recommendations_list = [(user, job) for user, jobs in all_recommendations.items() for job in jobs]
recommendations_df = pd.DataFrame(recommendations_list, columns=['resume_seq', 'job_id'])

# 결과 저장
recommendations_df.to_csv('ensemble_recommendations.csv', index=False, encoding='utf-8-sig
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load the application data
application_data = pd.read_csv('apply_train.csv')

def get_applied_jobs(user_id, application_data):
    return application_data[application_data['resume_seq'] == user_id]['recruitment_seq'].tolist()

def recommend_jobs_based_on_similarity(user_id, user_features, application_data, top_k_jobs=5):
    similar_jobs = []

    # Compute user similarity based on user_features
    user_similarity = cosine_similarity(user_features)
    user_idx = user_features.index.get_loc(user_id)

    # Get the jobs that this user applied for
    user_applied_jobs = get_applied_jobs(user_id, application_data)

    # Get similar users to this user
    similar_users = np.argsort(-user_similarity[user_idx])[:top_k_jobs]

    # Collect the jobs that similar users applied for (excluding the jobs that this user already applied for)
    for idx in similar_users:
        similar_user_id = user_features.index[idx]
        for job_id in get_applied_jobs(similar_user_id, application_data):
            if job_id not in user_applied_jobs:
                similar_jobs.append(job_id)

    # Return the top k jobs
    return similar_jobs[:top_k_jobs]

def get_all_recommendations(user_features, application_data):
    all_recommendations = {}
    all_users = application_data['resume_seq'].unique()
    
    for user_id in all_users:
        recommended_jobs = recommend_jobs_based_on_similarity(user_id, user_features, application_data)
        all_recommendations[user_id] = recommended_jobs
    
    return all_recommendations

all_user_recommendations = get_all_recommendations(user_features, application_data)

# Convert the results into a DataFrame and save to CSV
recommendations_df = pd.DataFrame(all_user_recommendations.items(), columns=['resume_seq', 'recommended_job_id'])
recommendations_df.to_csv('user_recommendations.csv', index=False, encoding='utf-8-sig')
```
