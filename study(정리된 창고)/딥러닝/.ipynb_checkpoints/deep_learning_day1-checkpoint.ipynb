```python
# 딥러닝 1일차
# 학습에 대한 전반적인 과정에 대해서
# 신경망
# 인간의 뉴런에 자극전달과정에서 아이디어를 착안해서 발생한 머신러닝 알고리즘이다.
# 시냅스를 통해서 다른 뉴런으로부터 자극을 전달받고 시냅스를 통해 다른 뉴런에게 자극을 전달한다.
# 신경망은 자극이 전달되는 과정을 그대로 표현한 알고리즘이다. 인공신경망에는 레이어와 퍼셉트론이 있다. 
# 레이어 = 신경세포 하나 단위, 뉴런 = 퍼셉트론, 시냅스 = 웨이트
# 일반적인 그림으로 네모가 퍼셉트론(뉴런),선을 웨이트(시냅스)라고 한다. 
# 퍼셉트론은 인풋(투입값), 웨이트(시냅스), 변환함수(뉴런 계산), 액티베이션(활성화 함수)으로 구성된다.
# 액티베이션(활성화 함수)은 다음 함수로 전달할 것인가? 말것인가?를 판단하는 역할을 수행한다.
# 액티베이션 함수는 시그모이드, 렐로, 스텝, 탄젠트, 리니어등 여러가지가 존재한다.
# 논문에서 보는 인공신경망 아웃풋은 다음과 같다.
# 만약 n의 숫자가 투입되어서 나온 값이 0보다 크다면 1을 출력하고 아닌경우 -1을 출력한다. 
# 예를 들어 2개가 입력되면 y = w1x1 - w2x2 - w0 라고 표현된다. 
# 이러한 모델을 만드는 법은 분류모델 만들때 쓰던 여러가지 (로지스틱, 의사결정 트리 등)를 활용한다. 
```

```python
# 다중 분류기 : 입력계층과 출력계층 사이에 소프트맥스가 있으면 다중분류기이다.
# 다중분류기에서 출력이 된 값들(로직)이 소프트맥스로 들어가면 확률로 바뀐다. 
# 소프트맥스는 각각의 값들을 확률로 바꿔준다. 가장 높은 확률로 예측을 해주는 것이다.
# 다중분류기에 중간에 히든 레이어를 추가한게 신경망이다.
# 얕은 신경망 : 입력계층과 출력계층 사이의 히든계층이 1개라면 얕은 신경망이다.
# 깊은 신경망 : 입력계층과 출력계층 사이의 히든계층이 2개 이상이라면 신경망이다.
# 신경망은 지도학습에 해당한다. 
```

```python
# 기존에는 입력데이터가 들어오면 바로 분류기로 전달되고 출력되는 케이스였다.
# 시그모이드 함수가 확인해서 0부터 1사이의 값을 가지고 0 또는 1로 출력했다. 기본 분류모델이다. 
# 아웃풋이 더 추가되었다. 
# 히든 레이어의 단위 한개 한개를 노드라고 하고 뉴런이라고도 한다. 
# 신경망은 구조가 다르다. 모든 입력에 대해 모든 노드(뉴런)에 연결된다. 
# 인풋이 3이고 노드가 4개면 12개의 인풋이 들어간다. 
# 그리고 b가 추가되어 인풋과 히든레이어 사이에는 13개의 퍼러미터가 존재한다.
# 아웃풋 노드에는 가중치가 존재한다. 아웃풋에 b가 가중된다. 
# 아웃풋은 시그모이드로 들어가서 0~1사이에서 판단하고 아웃풋을 한다. 
```

```python
# fully connected layers(완전연결신경망)
# 케라스와 텐서플로우로 하는 작업은 레이어에 있는 모든 노드를 연결하는 것이 첫번째다.
# 신경망을 공부하는 목적은 회귀와 분류작업을 위함이다.
# 모델의 성능을 올리기 위해서 여려 노력들이 있다. 그 부분도 수업에서 다룰 것이다.
# 딥러닝은 최근까지 인공지능에 가장 의미있게 발전된 분야이다. 
# 특히 자연어처리나 이미지 인식파트, 게임, 바둑 등 인간의 수준에 근접하거나 능가한 영역들이다.
# 신경망에서는 히든계층이 들어간다. 히든계층의 차이를 알기 위해서는 히든계층이 무엇인지 알아야 한다.
```

```python
# 기본모양
# 입력값 : g=공부시간, m=몸무게, c=출석일수
# 입력값을 확인하고 시험에 합격할까 불합격할까를 조사하고자 한다.
# 이 데이터 가지고 공부시간과 몸무게와 출석일수가 입력되었을때 시험에 합격하는지를 알아보고자 한다.
# 결과값을 y 라고 기준해보자. 
# g = 5 , m = 60 , c = 10 , y = 합격
# 여기서 어느 요인이 어느정도 중요하냐를 확인하는 것을 가중치라고 한다.
# 상식적으로 g와 c가 합격에 가중치를 많이 두고 m은 가중치가 적을 것이다.
# 공식은 gm + gc + mc + b = h(x)라고 예측값이 나올 것이다. 
# 이후 시그모이드로 전송한다. 0에서 1사이로 출력된다. 
# 이게 가장 기본적인 분류기를 만드는 과정이다. 
```

```python
# 히든계층
# 히든계층이 있으면 위의 기본모양에서 h(x)가 출력되는 것이 아니다. 
# 히든계층은 일종의 추상화된 특징(feature)이라고 할 수 있다. 
# 히든계층1 : 성실성, 히든계층2 : 근면성, 히든계층3 : 건강, 히든계층4 : 지식
# 성실성은 공부시간, 출석일수가 영향을 받을 것이다.
# 근면성은 출석일수에 영향을 받을 것이다.
# 건강은 몸무게에 영향을 받을 것이다.
# 지식은 공부시간, 출석일수 뿐 아니라 가중치(b)가 존재할 것이다.
# 이 4가지 특징으로 정리가 되었지만 합격, 불합격 요소를 결정하는 것에도 가중치가 다를 것이다.
# 시험으로 기준을 한다면 히든계층 4 - 2 - 1 - 3 순으로 가중치가 설정될 것이다.
# 요소들을 가중시킨 다음 시그모이드에 대입한다.
# 합격이면 1, 불합격이면 0 이렇게 출력될 것이다.
# 신경망을 쓴다는 이야기는 추상화된 특징을 둬서 인풋을 처리한다는 것이다. 
# 물론 머신러닝보다 성능이 더 떨어질 가능성도 존재한다. 그렇다면 어째서 신경망을 쓰는 것인가?
```

```python
# 히든계층을 쓰는 이유
# 중간에 히든계층을 두는가? 어떤 경우에 두는가?
# 만약 얼굴을 분류하는 모델을 만든다고 하자. 2차원 구조로 10x10이 있다고 하자.
# 플래트닝(flattening,평탄화)를 시행해서 위에서 아래로 이어서 100개짜리 1줄로 만든다. 
# 1차원 구조의 100차원으로 구성된 선이 만들어졌다. 차원마다 숫자가 다르다. 
# 로지스틱 회귀로 이미지를 분류한다고 해보자. 사람인지 강아지인지 판단하는 모델을 만든다고 하자.
# 이제 인풋은 100개이다. w1x1+ ... w100x100+b가 h(x)일 것이다. 
# 훈련과정을 확인해야 한다. 이 값들에 따라서 시그모이드 결과가 달라진다. b조차도 여기서 매우 중요하다.
# 학습과정은 다음과 같다.
# x1~x100까지는 이미 만들어져있다. 코스트가 가장 낮은 지점을 찾는 법은 머신러닝에서 배웠다.
# 여기서 x를 바꾸는 순간 데이터 조작이다. 가중치인 w와 b가 바뀌어야 한다. 
# x가 100이고 y가 300라고 하자. y = w*x라고 하자.  그러면 w는 3이 되어야 한다. w가 2라면 100만큼 애러가 난다.
# x1 = 100, x2 = 50, y = 300일때 y = x1w1 + x2+w2이다. 문제는 w1는 w2보다 훨씬 큰 영향을 끼친다. 
# w1가 1 변하면 100이 움직이고 w2가 1 변하면 50이 움직인다. 천지차이다. 
# 애러를 줄이기 위해서는 w와 b를 바꿔야 한다. x는 바꿀 수 없다.
# 그래서 미분이 필요하다. w들이 어느정도 영향을 끼치는지 확인해야 하기 때문이다.
# 실제값 y와 예측값 y의 차이점만큼 애러라고 했었다. 
# 여기서 w는 영향력이 각기 다르다. 그래서 w를 돌려보고 각기 어느정도의 영향력을 가졌는지 확인한다. 

# b값을 추가해보자. b는 x에 영향을 받지 않는 가중치이다. 
# x가 100이고 y가 3000이라고 하자. y = w*x+b라고 하자.
# 만약 w가 10이고 b가 500이면 1000+500으로 1500이다. y는 3000이니 1500만큼 애러가 발생했다. 

# 즉 합성함수의 미분이다. 애러에 대해서 각각의 데이터가 끼친 영향력을 보는 것이다.
# 또는 역 방향으로 전파해가면서 업데이트한다고 해서 백프라퍼게이션이(역전파)라고도 한다. 
# 애러가 발생한 만큼 백프라퍼게이션을 하고 웨이트를 바꾼다. 다시 테스트하고 포워드프라퍼게이션(순전파)을 한다. 이걸 반복한다.
# 사람사진 -포워드프라퍼게이션 -  개로 분류(애러) - 백프라퍼게이션(애러수정) - 다시 시도(포워드 프라퍼게이션)
# 이런 식으로 반복하는 것이다. 
# 데이터가 1000개일때 1000개의 데이터를 다 학습하면 1에폭이라고 한다.
# 2에폭은 다시 1000개의 데이터를 읽어서 다 하는 것이다. 
# 이런 식으로 시행착오를 겪으면서 판단한다. 1000에폭은 1000번 반복하는 것이다.
# 즉 수련을 하는 것이다. 이걸 학습이라고 한다. 
```

```python
# batch-size(배치사이즈)
# 배치사이즈가 몇 개의 데이터를 가지고 학습을 하느냐를 말한다. 즉 몇 단위로 한번에 업데이트를 하느냐를 의미한다.
# 즉 배치사이즈가 100이면 100건의 데이터도 한번에 읽어들여서 한번에 연산해버린다. 
# 백프라퍼게이션 다음에도 100건을 읽어서 연산해버린다. 
# 즉 배치사이즈 10은 데이터 1000건을 10개당 1뭉텅이로 해서 100번 계산하는 것이다.
# 데이터가 1000 개라고 하자. 배치사이즈를 3으로 하자. 총 333번을 돌린다. 
# 마지막 남은 1000번째는 1번과 2번을 포함해서 돌아간다.
```

```python
# 준비물 : 욜로(교육과정)으로 개인을 인식하는 것을 만들 것이다. 자신의 사진파일을 준비하자. 
```

```python
# 케라스 시퀸셜 클래스
# 입력이 3개라고 하자. 
# 입력 특징 : 공부시간, 과제제출횟수, 출석일수
# 출력 : 시험점수
# 케라스에서 제공하는 딥러닝 프레임웍을 이용해서 점수 예측을 해보자. 
# 입력은 3차원이고 출력은 1차원이다. 
# 속성을 먼저 지정해야 한다.
from tensorflow import keras # 케라스 호출 메모리에 케라스를 적재했다. 
from tensorflow.keras import layers # 케라스 레이어스 호출 케라스 내부에 레이어스를 호출해서 메모리에 적재했다.
```

```python
model = keras.Sequential([layers.Dense(units=1,input_shape=[3])])#3개의 값이 투입될 것이고 출력차원은 1개다.
# 즉 3개의 차원을 투입해서 1개의 차원으로 반환받는 것이다.
# 시퀸셜은 모듈이다. 히든계층을 중간에 나열한다는 뜻이다.  
# 덴스는 전체 연결을 말한다. fully connected layers(완전연결신경망)을 말한다.
# 
#              덴스
# 입력 ------| 히든
#            | 히든
# 입력 ------|--------- 출력      
#            | 히든
# 입력 ------| 히든

# 사실상 덴스가 히든계층이다. 
# 케라스에서 히든계층을 추가하는 방법은 덴스말고도 여러개가 존재한다. 
```

```python
# 컴퓨터에는 프레임워크라는 용어가 쓰인다. 대체로 딥러닝 프레임워크, 웹 프레임워크같이 앞에 단어가 들어간다.
# 프레임워크는 뼈대를 말한다. 어떠한 일을 처리하기 위한 구조화된 틀을 말한다.
# 맨 바닥에서 집을 짓기보다는 집을 구성하는 기반이 존재한다면 조립하듯이 건물을 지을 수 있을 것이다. 
# 수행하는 작업을 빠르고 안정적이고 쉽고 적은 비용으로 개발할 수 있도록 해주는 라이브러리 집합체이다. 
# 판다스는 데이터 분석 도구이다. 넘파이, 맷플랏립도 도구다. 
# 딥러닝 프레임워크는 딥러닝을 하기 위한 기반을 제시한다. 미분연산도 알아서 제공해준다.
# 만약 이것이 없으면 수학자가 아닌 이상 딥러닝을 다루기 매우 힘들다.
# 텐서플로, 파이토치, 까페, 케라스 등이 있다. 
# 케라스는 텐서플로와 별개였으나 텐서플로로 병합되었다. 텐서플로는 일반인이 다루기 어렵다. 
# 하지만 케라스는 상대적으로 인간친화적이다. 텐서플로를 이용하기 위해 케라스를 사용하는 것이다. 
```

```python
# 딥 뉴럴 네트워크는 복잡한 종류의 데이터를 가지고 관계를 학습할 수 있는 신경망을 구성할 수 있다.
# 복잡한 관계를 모델링해야하는데 이걸 설계하기 위해 단순한 신경망이 아닌 복잡한 신경망이 필요하다.
# 그래서 층으로 이를 나타내게 된다.
```

```python
# 렐루 
# ReLu = max(0,wx+b)
# 둘 중에 큰 값을 취하라는 뜻이다. 0과 예측모델값을 비교해서 큰 값을 출력하는 것이다.
# 즉 예측모델이 -값을 예측하면 0이 된다. 
# 시그모이드 기호는 원 안에 s자가 있고 렐루는 9시 10분 을 가리키는 모양의 기호로 나온다. 
# 렐루가 신경망에서 어느 위치에서 사용되는지에 대해 설명을 해볼 것이다. 
# 히든계층이 2개의 층 이상이 있다고 생각해보자. 

#              히든1    히든2
# 입력 ------| 히든 --/ 히든
#            | 히든 --/ 히든
# 입력 ------|-히든---/ 히든 ---- 출력      
#            | 히든 --/ 히든
# 입력 ------| 히든 --/ 히든
# 
# 이런 형식으로 배치가 되어있다고 하자.
# 히든1에서 받은 결과는 바로 히든 2로 넘어가지 않는다. 
# 활성화함수는 임계값을 넘어가면 1로 보내고 아니면 0으로 보낸다.
# 히든1 노드들에서 활성화함수를 거쳐서 히든2로 들어간다. 노드마다 전부 활성화함수가 있다.
# 활성화함수가 계산을 마치고 다음으로 넘긴다. 

#              히든1                히든2
# 입력 ------| 히든 --활성화함수--/ 히든--활성화함수--|
#            | 히든 --활성화함수--/ 히든--활성화함수--|
# 입력 ------|-히든---활성화함수--/ 히든--활성화함수--|---- 출력(합이 1인 출력값)      
#            | 히든 --활성화함수--/ 히든--활성화함수--|
# 입력 ------| 히든 --활성화함수--/ 히든--활성화함수--|
#
# 활성화함수는 0~1사이의 값을 구해서 히든 2로 보내고 히든2에서 다시 활성화함수를 통해 출력절차로 간다. 
# 시그모이드는 중간에서 잘 안쓰고 렐루를 많이 쓴다. 
# 왜냐하면 백프라퍼게이션을 진행해야하기 때문이다. 애러가 발생했을때 역전파를 해서 업데이트를 해야한다.
# 근데 깊은 신경망은 합성함수 같은 형식이다. 오류가 발생했을때 히든1과 히든2의 결과값을 곱해서 오류체크를 한다.
# 문제가 마이너스는 곱셈에 노이즈를 매우 많이 발생시킨다. (vanishing. 소실문제) 
# 시그모이드는 0~1사이이다. 곱해봐야 0~1사이에서 나온다. 
# 처음에는 시그모이드를 쓰다가 딥러닝을 쓰는 순간 학습이 아예 안되는 부작용이 생겼다. 
# 물론 렐루도 단점이 있어서 0 이하의 값은 전부 0으로 처리해서 정밀성이 떨어졌다.
# 그래서 리키렐루가 만들어졌다. 8시 10분을 가리키는 모양의 리키렐루는 음수일때도 학습이 이루어지게 설계되어있다. 
```

```python
#              히든1                히든2
# 입력 ------| 히든 --활성화함수--/ 히든--활성화함수|
#            | 히든 --활성화함수--/ ----------------|
#            | -------------------/ 히든--활성화함수|---- 출력      
#            | 히든 --활성화함수--/ ----------------|
# 입력 ------| 히든 --활성화함수--/ 히든--활성화함수|

# 이렇게 표현해보자. 
model = keras.Sequential([
layers.Dense(units=4, input_shape=[2], activation='relu'), # 히든으로 출력하는 것이기에 인풋2 출력4이다. 활성함수는 렐루이다.
layers.Dense(units=3, activation='relu'), # 컴퓨터는 똑똑해서 중간값을 생략해도 4가 인풋인걸 알아듣는다.    
layers.Dense(units=1, activation='sigmoid') # 마지막 값은 시그모이드로 출력해보자. 
])
```

```python
model.summary() #내가 만든 모델의 구조이다. 
# 맨 위의 퍼러미터가 12이다. 2x4로 8개에서 바이어스 변수 4를 더해서 12이다. 
# 두번째 퍼러미터는 4x3+3으로 15개이다. 
# 아웃풋 퍼러미터는 3x1+1이다. 
# 31개의 학습이 필요하다. 그리고 그리고 한개당 4바이트로 퍼러미터가 31개가 나온다.
# gpt4는 억단위의 퍼러미터가 존재한다.
```

```python
# 웨이트의 값.
#              히든1                히든2
# 입력 ------| 히든 --활성화함수--/ 히든--활성화함수|
#            | 히든 --활성화함수--/ ----------------|
#            | -------------------/ 히든--활성화함수|---- 출력      
#            | 히든 --활성화함수--/ ----------------|
# 입력 ------| 히든 --활성화함수--/ 히든--활성화함수|

# 위 그림에서 -로 표시한 것이 웨이트이다. 입력계층과 히든계층과 출력 사이에 --로 표시되어있다.
# 웨이트에 할당하는 값은 랜덤이다.
# 즉 처음에는 아무것도 모르는 상태로 시작한다.
# 처음에는 그냥 랜덤한 값으로 만든 가상의 모델에 지나지 않는다. 하지만 학습을 거듭하면서 굉장히 똑똑해진다. 
# 훈련데이터를 학습하여 모델을 만들것이다. 
# 각각의 데이터는 입력데이터와 출력데이터가 함께 있는 채로 구성되어있을 것이다. 
# 입력에 해당되는 특징과 출력에 해당되는 특징을 값 사이에 식으로 표현하게 될 것이다.
# 그것이 바로 모델이 되는 것이다.
# 훈련의 의미는 x변수(입력변수)로 최종적으로 얻고 싶은 값이 실제값과 유사하도록 만드는 바이오스, 웨이트 조정과정이다.
# 즉 크리크 수정이다. 예측모델을 만든다는 것은 가중치를 조정해서 최적의 가중치를 찾아내는 것이다. 그게 훈련이다.
# 우리의 모델이 최종적으로 훈련이 이루어지기 위해서는 
# 가중치가 입력데이터의 특징들과 출력값 사이에 식으로 잘 표현할 수 있도록 해야한다. 
```

```python
# 좋은 딥러닝 모델이란?
# 좋은 모델의 기준은 코스트 함수로 알 수 있다. 로스함수(손실함수)라고도 한다.
# 로스함수의 값이 가장 작게 나오는 w와 b를 훈련을 통해서 알게 되는 것이다. 
# 가중치를 변경하는 것은 컴퓨터(텐서플로/케라스)가 한다. 
# 딥러닝 프레임워크는 가중치를 바꾸는 법을 알고있다. 우리는 따로 세부적인 옵션만 주면 적절하게 가중치조정을 하게된다.
# 가중치를 변경하는 법은 여러개가 있다. (그레디언트 디센트 알고리즘-기초적인 방법, 아담 옵티마이저-많이 씀, rms prof, momentum 등)
# 알고리즘은 상당히 많이 나오고 있다. 현재에는 아담 옵티마이저가 대세이다. 머신러닝의 랜덤포레스트 포지션이다.
# 그레디언트 디센트 알고리즘은 산에서 하산하는 과정과 비슷하다.
# 경사가 밑으로 내려가는 것을 확인하고 내려가는 쪽으로 천천히 가는 것이다. 
# 최적의 경로로 내려오는 것이 목적이다. 최적의 w와 b를 찾는 것이 목적이다. 
# 정답에 가장 근접한 값을 찾아내기 위해 미분을 사용한다. 
# 옵티마이저도 이와 같다. 하나같이 기울기를 가지고 내려가는 방식으로 수행된다. 
# 반복적으로 훈련하여 조금씩 조금씩 기울기가 낮아지는 방향으로 업데이트 된다.
# 웨이트 수식 업데이트도 코스트의 영향력을 고려해서 업데이트 된다. 그렇게 애러를 최소화하면서 나아간다.
```

```python
# 테스트를 거듭할수록 기울기와 y절편이 바뀌어간다.
# 점점 시간이 지날수록 0에 가깝게 수렴해간다. 
# 질문 : 잘 수행되었을때 코스트가 크게 나왔을 때 원인은 무엇인가?
# 대답 : 실제값과 예측값이 일치하면 0이다. 이런 경우는 거의 없다. 
# 모델의 주변의 데이터들이 있을 것이다. 완전 선형적인 데이터는 사실상 없다. 
# 선형성이 많이 없다면 코스트가 아무리 많이 떨어져도 더 이상 변화가 없을 수 있다. 그게 최적의 코스트이다.
# 러닝 메이트를 조절해서 학습을 잘 이루어지도록 해야한다. 이래도 코스트가 높다면 낮추긴 힘들다. 
# 미니배치가 이루어지면서 가중치가 업데이트 된다. 
# 아담이 범용적인 옵티마이저이다. 
```

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
from tensorflow.keras.models import Sequential # 시퀸셜도 바로 쓰자. 
from tensorflow.keras.layers import Dense # 걍 덴스를 바로 쓰자. 
```

```python
# 오랜만에 사용한다. 당뇨병 데이터이다.
# 이진분류를 할 때에는 시그모이드를 사용한다.
x = df.iloc[:,0:8] # 훈련데이터다.
y = df.iloc[:,8] # 테스트데이터이다. 
```

```python
model = Sequential()
```

```python
model.add(Dense(12,input_dim=8,activation='relu')) # dim은 바로 숫자를 쓰게 해준다. 
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
# 8 x 12 + 12 = 108 이건 이해갈 것이다. 
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=10,batch_size=5)
# 정확도는 68%이다. 아직은 만족스럽진 않다. 
```

```python
df = pd.read_csv('iris3.csv')
df # 종을 찾는 것은 소프트맥스를 사용하면 될 것이다. 
```

```python
x = df.iloc[:,0:4]
y = df.iloc[:,4]
```

```python
y = pd.get_dummies(y)
y # 순서가 없는 데이터타입의 경우에는 원핫인코딩이 편하다. 
```

```python
model=Sequential()
```

```python
model.add(Dense(16,input_dim=4,activation='relu')) # 4개의 차원을 투입해야 한다.
model.add(Dense(8,activation='relu')) 
model.add(Dense(3,activation='sigmoid'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=40,batch_size=5) # 정확도가 96%에서 98%가량 나온다. 정확도가 나쁘진 않다. 
```

```python
# 만약 한개의 데이터를 제공하고 정답을 받고자 한다면?
model.predict(np.array([[5.9,3.0,5.1,1.8]])) # 각각의 확률을 출력하고 최고확률을 답으로 출력한다. 
```
