```python
df = pd.read_csv('zoo/zoo.data', header=None)                                            # 데이터 로드
index_columns = ['animal_name', 'hair', 'feathers', 'eggs', 'milk', 'airborne',          # 인덱스 이름 붙이기
                 'aquatic', 'predator', 'toothed', 'backbone', 'breathes', 'venomous',
                 'fins', 'legs', 'tail', 'domestic', 'catsize', 'type']

df.columns = index_columns                                                              # 열 이름 할당하기 
```

```python
df.drop(['animal_name'],axis=1,inplace=True)                                            # 동물이름 드롭하기
```

```python
df
```

```python
df2= pd.DataFrame(np.tile(np.array(df),(5,1)))                                          # 동물이름 복사(np.tile 이용하여 복사)
df2.columns = index_columns[1:]                                                         # 동물 이름 제외하고 데이터이름 할당하기
```

```python
df2.columns = index_columns[1:]                                                         # 동물 이름 제외하고 데이터이름 할당하기
legs = pd.get_dummies(df2['legs'],prefix='legs_num')                                    # 동물 다리개수 원 핫 인코딩 하기
legs = legs.astype('int')                                                               # 타입을 숫자로 하여 0,1로 구분하기
df3 = pd.concat([df2,legs],axis=1)                                                      # 열 방향으로 다리개수 합하기
df3.drop('legs',axis=1,inplace=True)                                                    # 다리 항목 드랍하기
df3.info()
```

```python
df3.drop('legs',axis=1,inplace=True)                                                    # 다리 항목 드랍하기
from sklearn.model_selection import train_test_split                                    # 훈련/테스트셋 분리

xtrain, xtest,ytrain, ytest=train_test_split(                                           # 트레인데이터, 테스트데이터 분리결과
    df3.drop(['type'],axis=1),df3.type,                                                 # 정답 데이터 드롭한것이 트레인, 정답 데이터는 답지
    test_size=0.3,random_state= 1122)                                                   # 테스트용으로는 30% 사용, 랜덤값은 1122로 고정  
ytest2 = ytest                                                                          # 별도로 테스트데이터 빼기
```

```python
import tensorflow as tf                                                                 # 텐서플로 호출 

from tensorflow.keras.models import Sequential                                          # 케라스 모델 호출
from tensorflow.keras.layers import Dense                                               # dense호출                             
```

```python
ytrain = tf.keras.utils.to_categorical(ytrain, num_classes=8)                           # 훈련답지 클래스 변경
```

```python
xtrain
```

```python
ytrain = tf.keras.utils.to_categorical(ytrain, num_classes=8)                           # 훈련답지 클래스 변경
model = tf.keras.Sequential()                                                           # 케라스 모델 제작
model.add(tf.keras.layers.Dense(64, input_shape=(21,)))                                 # 특징 21차원 할당.64차원 아웃풋
model.add(tf.keras.layers.Dense(64, activation='relu'))                                 # 렐루로 계산하기
model.add(tf.keras.layers.Dense(8, activation='softmax'))                               # 8차원 소프트맥스 출력
model.summary()                                
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),                   # 모델 컴파일. 옵티마이저 아담
              loss='categorical_crossentropy',                                          # 다중 정답은 카테고리 교차엔트로피
              metrics=['accuracy'])                                                     # 정확도 기준
```

```python
hist = model.fit(xtrain, ytrain, epochs=30, validation_split=0.3)                       # 에폭 30으로 테스트 시작
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),                   # 모델 컴파일. 옵티마이저 아담
              loss='categorical_crossentropy',                                          # 다중 정답은 카테고리 교차엔트로피
              metrics=['accuracy'])                                                     # 정확도 기준
model.evaluate(xtest, ytest)                                                            # 모델평가
```

```python
import torch                                                                            # 파이토치 불러오기 
import torch.nn as nn                                                                   # 뉴럴네트워크 호출
import torch.nn.functional as F                                                         # 뉴럴네트워크 함수호출
import torch.optim as optim                                                             # 토치 옵티마이저 호출
```

```python
torch.manual_seed(1122)  
```

```python
# 모델 및 데이터를 PyTorch Tensor로 변환

x_train_numpy = xtrain.values                                                           # 데이터프레임을 NumPy 배열로 변환
y_train_numpy = ytrain                                                                  # 데이터프레임을 NumPy 배열로 변환

x_train_tensor = torch.tensor(x_train_numpy, dtype=torch.float32)                       # NumPy 배열을 PyTorch Tensor로 변환
y_train_tensor = torch.tensor(y_train_numpy, dtype=torch.float32)                       # NumPy 배열을 PyTorch Tensor로 변환

model = torch.nn.Linear(21, 8)                                                          # 모델투입 21,8 투입
criterion = torch.nn.MSELoss()                                                          # mse를 손실함수로 지정
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)                                # 옵티마이저 sgd로 지정

for epoch in range(2000):                                                               # 2000에폭 설정
    hx = model(x_train_tensor)                                                          # 모델에 훈련 데이터 투입
    cost = criterion(hx, y_train_tensor)                                                # 코스트 계산
    optimizer.zero_grad()                                                               # 옵티마이저 설정
    cost.backward()                                                                     # 확인필요 
    optimizer.step()                                                                    # 확인필요

    if epoch % 100 == 0:
        print(epoch, cost.item())                                                       # 100에폭마다 코스트 확인
```
